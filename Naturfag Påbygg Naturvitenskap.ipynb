```python
import locale
import tkinter as tk
from tkinter import messagebox, ttk
import random

# Lokalitet (for å unngå feil med ttkbootstrap)
try:
    locale.setlocale(locale.LC_ALL, 'en_US.UTF-8')
except locale.Error:
    locale.setlocale(locale.LC_ALL, '')

from ttkbootstrap import Style
locale.setlocale(locale.LC_ALL, '')

# Opprinnelige quiz-data (lagres separat for reset)
original_quiz_data = [
    {
        "question": "Hva menes med at en hypotese skal kunne falsifiseres?",
        "options": ["A. At den må være sann", "B. At den må være populær", "C. At den må kunne motbevises gjennom eksperimenter", "D. At den må være logisk"],
        "answer": "C",
        "explanation": "En hypotese må kunne motbevises for å være vitenskapelig; det vil si at det må være mulig å teste den på en måte som kan vise at den er feil."
    },
    {
        "question": "Hva menes med kildekritikk?",
        "options": ["A. Å lese mange kilder", "B. Å stole på den første kilden man finner", "C. Å vurdere hvor troverdig og relevant en kilde er", "D. Å skrive kilder med korrekt format"],
        "answer": "C",
        "explanation": "Kildekritikk handler om å vurdere hvor pålitelig, relevant og objektiv informasjonen er, særlig viktig i digitale medier."
    },
    {
        "question": "Hva er en variabel i naturvitenskapelige eksperimenter?",
        "options": ["A. En konstant faktor", "B. En tilfeldig målefeil", "C. En faktor som kan endres eller måles", "D. En naturlov"],
        "answer": "C",
        "explanation": "En variabel er noe som kan påvirkes eller måles i et eksperiment, som temperatur, tid eller mengde lys."
    },
    {
        "question": "Hvorfor er bruken av modeller viktig i naturvitenskapen?",
        "options": ["A. Fordi det er billigere", "B. Fordi det ser bedre ut i rapporter", "C. Fordi modeller forenkler komplekse fenomener og gjør dem lettere å forstå", "D. Fordi modeller alltid er helt nøyaktige"],
        "answer": "C",
        "explanation": "Modeller hjelper oss å forstå og forutsi komplekse fenomener ved å forenkle virkeligheten."
    },
    {
        "question": "Hva er hensikten med datasimuleringer i naturfag?",
        "options": ["A. Å unngå eksperimenter", "B. Å kontrollere medieinnhold", "C. Å teste scenarier digitalt som er vanskelig å teste i virkeligheten", "D. Å skrive bedre rapporter"],
        "answer": "C",
        "explanation": "Datasimuleringer gjør det mulig å teste hypoteser og scenarier som ville vært for dyre, farlige eller upraktiske å teste i virkeligheten."
    },
    {
        "question": "Hva bør en naturfaglig rapport inneholde?",
        "options": ["A. Bilder og kilder", "B. Teorier og hypoteser", "C. Innledning, hypotese, metode, resultater, diskusjon og konklusjon", "D. En liste over modeller"],
        "answer": "C",
        "explanation": "En naturfaglig rapport skal være strukturert med innledning, hypotese, metode, resultater, diskusjon og konklusjon for å dokumentere eksperimentet tydelig."
    },
    {
        "question": "Hva omfatter naturvitenskap ifølge boka?",
        "options": ["A. Bare kunnskap om naturen", "B. Bare vitenskapelig metode", "C. Kunnskap og arbeid for å forstå naturen", "D. Teorier som ikke kan testes"],
        "answer": "C",
        "explanation": "Naturvitenskap omfatter både kunnskap om naturen og måten vi arbeider på for å forstå den."
    },
    {
        "question": "Hva betyr det at naturvitenskap er empirisk?",
        "options": ["A. Den bygger på følelser", "B. Den bygger på erfaring og observasjon", "C. Den er basert på tro", "D. Den er basert på gamle teorier"],
        "answer": "B",
        "explanation": "Empirisk betyr at kunnskapen er basert på erfaring, observasjon og eksperimenter."
    },
    {
        "question": "Hva er en hypotese?",
        "options": ["A. En tilfeldig antakelse", "B. En testbar antakelse om en observasjon", "C. En konklusjon fra et eksperiment", "D. En kjent teori"],
        "answer": "B",
        "explanation": "En hypotese er en testbar antakelse som forsøker å forklare en observasjon eller et fenomen."
    },
    {
        "question": "Hva er en feilkilde i et eksperiment?",
        "options": ["A. En feil i teorien", "B. En feil i hypotesen", "C. En faktor som kan påvirke resultatet", "D. En feil i boka"],
        "answer": "C",
        "explanation": "Feilkilder er faktorer som kan påvirke resultatene og gjøre dem mindre pålitelige."
    },
    {
        "question": "Hvordan kan man redusere usikkerhet i målinger?",
        "options": ["A. Gjøre færre målinger", "B. Bruke tilfeldige personer", "C. Gjenta forsøket mange ganger", "D. Måle uten å bruke klokke"],
        "answer": "C",
        "explanation": "Flere målinger gir et mer pålitelig gjennomsnitt og reduserer usikkerhet."
    },
    {
        "question": "Hva er generalisering?",
        "options": ["A. Å bruke samme metode i alle fag", "B. Å trekke konklusjoner fra ett forsøk til alle situasjoner", "C. Å bruke tilfeldige data", "D. Å ignorere usikkerhet"],
        "answer": "B",
        "explanation": "Generalisering handler om å vurdere om resultatene gjelder i andre situasjoner også."
    },
    {
        "question": "Hva er en modell i naturfag?",
        "options": ["A. En eksakt kopi av virkeligheten", "B. En forenklet fremstilling av virkeligheten", "C. En tilfeldig tegning", "D. En teori som ikke kan testes"],
        "answer": "B",
        "explanation": "En modell er en forenklet fremstilling som hjelper oss å forstå komplekse fenomener."
    },
    {
        "question": "Hvorfor er det viktig å diskutere usikkerhet i forsøk?",
        "options": ["A. For å gjøre rapporten lengre", "B. For å vise at man ikke kan stole på resultatene", "C. For å øke troverdigheten", "D. For å unngå å gjøre forsøket igjen"],
        "answer": "C",
        "explanation": "Diskusjon om usikkerhet øker troverdigheten til resultatene og forskerne."
    },
    {
        "question": "Hva er en observasjon?",
        "options": ["A. En teori", "B. En måling eller sanseinntrykk", "C. En hypotese", "D. En konklusjon"],
        "answer": "B",
        "explanation": "Observasjoner er det vi ser, hører, lukter eller måler i et forsøk."
    },
    {
        "question": "Hva er en teori i naturvitenskap?",
        "options": ["A. En gjetning", "B. En testbar hypotese", "C. En godt dokumentert forklaring", "D. En modell"],
        "answer": "C",
        "explanation": "En teori er en godt dokumentert og testet forklaring på et fenomen."
    },
    {
        "question": "Hva er en stamform?",
        "options": ["A. En utdødd art", "B. En felles forfader til flere arter", "C. En modell", "D. En hypotese"],
        "answer": "B",
        "explanation": "En stamform er en felles forfader som to eller flere arter har utviklet seg fra."
    },
    {
        "question": "Hva er en feilkilde ved manuell tidtaking?",
        "options": ["A. Klokken er for nøyaktig", "B. Reaksjonstiden varierer", "C. Det er for mange personer", "D. Det er for få målinger"],
        "answer": "B",
        "explanation": "Reaksjonstiden til den som måler kan føre til feil i tidtakingen."
    },
    {
        "question": "Hva er en måte å undersøke skjermtid og søvnkvalitet på?",
        "options": ["A. Spørre om favorittprogrammer", "B. Måle søvn med klokke", "C. Sammenligne grupper med ulik skjermtid", "D. Se på mobilbruk i helgene"],
        "answer": "C",
        "explanation": "Man kan sammenligne grupper med ulik skjermtid og se på forskjeller i søvnkvalitet."
    },
    {
        "question": "Hva er en svakhet ved plastikkjuletrær som modell for grantrær?",
        "options": ["A. De er for store", "B. De lukter ikke gran", "C. De er ekte trær", "D. De har kongler"],
        "answer": "B",
        "explanation": "Plastikktrær lukter ikke gran og vokser ikke, og er derfor ikke gode modeller for levende grantrær."
    },
    {
        "question": "Hva er en svakhet ved atommodellen med kuler?",
        "options": ["A. Den viser ikke elektroner", "B. Den viser ikke farge", "C. Den viser ikke bindinger", "D. Alle alternativene"],
        "answer": "D",
        "explanation": "Kulemodellen viser ikke elektroner, bindinger eller hvordan atomene faktisk ser ut."
    },
    {
        "question": "Hva er oppdragsforskning?",
        "options": ["A. Forskning gjort av elever", "B. Forskning som er betalt av en tredjepart", "C. Forskning som er gratis", "D. Forskning som ikke publiseres"],
        "answer": "B",
        "explanation": "Oppdragsforskning er forskning som er bestilt og betalt av for eksempel myndigheter eller firmaer."
    },
    {
        "question": "Hva kan VR-teknologi brukes til i naturfag?",
        "options": ["A. Å spille spill", "B. Å modellere molekyler i 3D", "C. Å lage videoer", "D. Å skrive rapporter"],
        "answer": "B",
        "explanation": "VR-teknologi kan brukes til å modellere komplekse molekyler i 3D."
    },
    {
        "question": "Hva skjer med gaupebestanden når harebestanden synker?",
        "options": ["A. Den øker", "B. Den holder seg stabil", "C. Den synker", "D. Den forsvinner"],
        "answer": "C",
        "explanation": "Når det blir færre harer, får gaupene mindre mat og bestanden synker."
    },
    {
        "question": "Hva er en viktig grunn til å bruke modeller i naturfag?",
        "options": ["A. For å gjøre ting enklere å forstå", "B. For å unngå eksperimenter", "C. For å lage kunst", "D. For å unngå feil"],
        "answer": "A",
        "explanation": "Modeller gjør det lettere å forstå komplekse fenomener."
    }
]

# Initialiser GUI
root = tk.Tk()
root.title("Quiz App")
root.geometry("600x550")
style = Style(theme="flatly")
style.configure("TLabel", font=("Helvetica", 16))
style.configure("TButton", font=("Helvetica", 14))

# Globale variabler
score = 0
current_question = 0
choice_btns = []
quiz_data = []
user_answers = []

# Widgets
qs_label = ttk.Label(root, anchor="center", wraplength=500, padding=10)
qs_label.pack(pady=10)

choices_frame = ttk.Frame(root)
choices_frame.pack(pady=10, fill=tk.X)

feedback_label = ttk.Label(root, anchor="center", padding=10, wraplength=500, justify="left")
feedback_label.pack(pady=10)

score_label = ttk.Label(root, text="", anchor="center", padding=10)
score_label.pack(pady=10)

nav_frame = ttk.Frame(root)
nav_frame.pack(pady=10)

prev_btn = ttk.Button(nav_frame, text="Tilbake", command=lambda: previous_question(), state="disabled")
prev_btn.pack(side="left", padx=10)

next_btn = ttk.Button(nav_frame, text="Neste", command=lambda: next_question(), state="disabled")
next_btn.pack(side="left", padx=10)

retry_btn = ttk.Button(root, text="Prøv igjen", command=lambda: restart_quiz(), state="disabled")
retry_btn.pack(pady=10)

progress = ttk.Progressbar(root, orient="horizontal", length=400, mode="determinate")
progress.pack(pady=10)

# --- Funksjoner ---

def restart_quiz():
    global score, current_question, choice_btns, quiz_data, user_answers
    score = 0
    current_question = 0
    quiz_data = original_quiz_data.copy()
    random.shuffle(quiz_data)
    user_answers = [None] * len(quiz_data)
    progress.config(maximum=len(quiz_data), value=0)
    retry_btn.config(state="disabled")
    next_btn.config(state="disabled")
    prev_btn.config(state="disabled")
    show_question()

def show_question():
    global choice_btns
    for btn in choice_btns:
        btn.destroy()
    choice_btns = []

    question = quiz_data[current_question]
    qs_label.config(text=question["question"])
    options = question["options"]

    selected_index = user_answers[current_question]

    for i, option in enumerate(options):
        button = ttk.Button(
            choices_frame,
            text=option,
            command=lambda i=i: check_answer(i),
            style="TButton"
        )
        button.pack(pady=5, fill=tk.X)
        choice_btns.append(button)

    feedback_label.config(text="")
    score_label.config(text="Score: {}/{}".format(score, len(quiz_data)))
    prev_btn.config(state="normal" if current_question > 0 else "disabled")
    next_btn.config(state="disabled")
    retry_btn.config(state="disabled")

    if selected_index is not None:
        apply_feedback_colors(selected_index)
        next_btn.config(state="normal")

def check_answer(choice_index):
    global score
    question = quiz_data[current_question]
    selected_letter = chr(65 + choice_index)
    correct_letter = question["answer"]
    correct_index = ord(correct_letter) - 65
    previous = user_answers[current_question]

    # Juster score hvis bytter svar
    if previous is not None and chr(65 + previous) == correct_letter:
        score -= 1

    user_answers[current_question] = choice_index

    if selected_letter == correct_letter:
        score += 1
        feedback_label.config(text="Riktig!\n" + question['explanation'], foreground="green")
    else:
        feedback_label.config(
    text=f"Galt! Det riktige svaret er {correct_letter}.\n{question['explanation']}",
    foreground="red"
        )

    score_label.config(text="Score: {}/{}".format(score, len(quiz_data)))
    progress['value'] = current_question + 1

    apply_feedback_colors(choice_index)
    next_btn.config(state="normal")

def apply_feedback_colors(selected_index):
    question = quiz_data[current_question]
    correct_index = ord(question["answer"]) - 65

    for i, button in enumerate(choice_btns):
        button.config(state="disabled")
        if i == correct_index:
            button.config(style="success.TButton")
        elif i == selected_index:
            button.config(style="danger.TButton")
        else:
            button.config(style="secondary.TButton")

def next_question():
    global current_question
    if current_question < len(quiz_data) - 1:
        current_question += 1
        show_question()
    else:
        messagebox.showinfo(
            "Quiz gjennomført",
            f"Du fikk {score} av {len(quiz_data)} riktige.\nTakk for at du deltok!"
        )
        retry_btn.config(state="normal")
        next_btn.config(state="disabled")
        prev_btn.config(state="disabled")

def previous_question():
    global current_question
    if current_question > 0:
        current_question -= 1
        show_question()

# Start første gang
restart_quiz()
root.mainloop()
```
