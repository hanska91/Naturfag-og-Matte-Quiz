```python
!pip install ttkbootstrap
import locale
import tkinter as tk
from tkinter import messagebox, ttk
import random

# Lokalitet (for å unngå feil med ttkbootstrap)
try:
    locale.setlocale(locale.LC_ALL, 'en_US.UTF-8')
except locale.Error:
    locale.setlocale(locale.LC_ALL, '')

from ttkbootstrap import Style
locale.setlocale(locale.LC_ALL, '')

# Opprinnelige quiz-data (lagres separat for reset)
original_quiz_data = [
        {
            "question": "Hvilket av følgende er en monosakkarid?",
            "options": ["A. Sukrose", "B. Laktose", "C. Glukose", "D. Stivelse"],
            "answer": "C",
            "explanation": "Glukose er en enkel sukkerart (monosakkarid) som er en viktig energikilde for kroppen."
        },
        {
            "question": "Hva er hovedfunksjonen til proteiner i kroppen?",
            "options": ["A. Energikilde", "B. Byggesteiner for celler og vev", "C. Lagring av genetisk informasjon", "D. Transport av oksygen"],
            "answer": "B",
            "explanation": "Proteiner fungerer som byggesteiner for celler og vev, og er essensielle for vekst og reparasjon av kroppen."
        },
        {
            "question": "Hvilket fettstoff er kjent som 'godt' fett?",
            "options": ["A. Mettet fett", "B. Transfett", "C. Umettet fett", "D. Kolesterol"],
            "answer": "C",
            "explanation": "Umettet fett, som finnes i fisk, nøtter og olivenolje, er kjent som 'godt' fett fordi det kan bidra til å redusere risikoen for hjertesykdom."
        },
        {
            "question": "Hvilket mineral er viktig for beinhelsen?",
            "options": ["A. Jern", "B. Kalsium", "C. Natrium", "D. Kalium"],
            "answer": "B",
            "explanation": "Kalsium er viktig for å opprettholde sterke bein og tenner, og spiller en rolle i muskel- og nervefunksjon."
        },
        {
            "question": "Hva er en funksjon av vann i kroppen?",
            "options": ["A. Energikilde", "B. Byggesteiner for celler", "C. Temperaturregulering", "D. Transport av oksygen"],
            "answer": "C",
            "explanation": "Vann hjelper med å regulere kroppstemperaturen gjennom svette og pust, og er essensielt for mange fysiologiske prosesser."
        },
        {
            "question": "Hvilket av følgende er en polysakkarid?",
            "options": ["A. Fruktose", "B. Maltose", "C. Cellulose", "D. Galaktose"],
            "answer": "C",
            "explanation": "Cellulose er en polysakkarid som finnes i plantecellevegger og er en viktig kilde til kostfiber."
        },
        {
            "question": "Hvilket sporstoff er viktig for skjoldbruskkjertelens funksjon?",
            "options": ["A. Sink", "B. Jod", "C. Selen", "D. Kobber"],
            "answer": "B",
            "explanation": "Jod er nødvendig for produksjon av skjoldbruskkjertelhormoner, som regulerer stoffskiftet."
        },
        {
            "question": "Hvilket av følgende er en disakkarid?",
            "options": ["A. Glukose", "B. Fruktose", "C. Laktose", "D. Stivelse"],
            "answer": "C",
            "explanation": "Laktose er en disakkarid som finnes i melk og består av glukose og galaktose."
        },
        {
            "question": "Hva er hovedfunksjonen til fett i kroppen?",
            "options": ["A. Energikilde", "B. Byggesteiner for celler", "C. Transport av oksygen", "D. Lagring av genetisk informasjon"],
            "answer": "A",
            "explanation": "Fett fungerer som en viktig energikilde og lagrer energi for senere bruk."
        },
        {
            "question": "Hvilket mineral er viktig for oksygentransport i blodet?",
            "options": ["A. Kalsium", "B. Jern", "C. Magnesium", "D. Sink"],
            "answer": "B",
            "explanation": "Jern er en komponent av hemoglobin, som transporterer oksygen i blodet."
        },
        {
            "question": "Hvilket vitamin er viktig for synet?",
            "options": ["A. Vitamin A", "B. Vitamin B12", "C. Vitamin C", "D. Vitamin D"],
            "answer": "A",
            "explanation": "Vitamin A er essensielt for synet og bidrar til å opprettholde en sunn netthinne."
        },
        {
            "question": "Hvilket vitamin er viktig for opptak av kalsium?",
            "options": ["A. Vitamin A", "B. Vitamin B12", "C. Vitamin C", "D. Vitamin D"],
            "answer": "D",
            "explanation": "Vitamin D hjelper kroppen med å absorbere kalsium, som er nødvendig for sunne bein."
        },
        {
            "question": "Hvilket vitamin er kjent for å styrke immunforsvaret?",
            "options": ["A. Vitamin A", "B. Vitamin B12", "C. Vitamin C", "D. Vitamin K"],
            "answer": "C",
            "explanation": "Vitamin C er kjent for å styrke immunforsvaret og beskytte mot infeksjoner."
        },
        {
            "question": "Hvilket vitamin er viktig for blodkoagulering?",
            "options": ["A. Vitamin A", "B. Vitamin B12", "C. Vitamin C", "D. Vitamin K"],
            "answer": "D",
            "explanation": "Vitamin K er nødvendig for blodkoagulering og hjelper med å forhindre blødninger."
        },
        {
            "question": "Hvilket vitamin er viktig for dannelse av røde blodceller?",
            "options": ["A. Vitamin A", "B. Vitamin B12", "C. Vitamin C", "D. Vitamin D"],
            "answer": "B",
            "explanation": "Vitamin B12 er nødvendig for produksjon av røde blodceller og forhindrer anemi."
        },
        {
            "question": "Hva er en viktig kilde til omega-3 fettsyrer?",
            "options": ["A. Rødt kjøtt", "B. Smør", "C. Fet fisk", "D. Hvitt brød"],
            "answer": "C",
            "explanation": "Fet fisk som laks og makrell er rike på omega-3 fettsyrer, som er gunstige for hjertehelsen."
        },
        {
            "question": "Hvilket næringsstoff er viktig for muskelvekst og reparasjon?",
            "options": ["A. Karbohydrater", "B. Fett", "C. Proteiner", "D. Vitaminer"],
            "answer": "C",
            "explanation": "Proteiner er essensielle for muskelvekst og reparasjon etter fysisk aktivitet."
        },
        {
            "question": "Hvilket vitamin kan kroppen produsere når huden utsettes for sollys?",
            "options": ["A. Vitamin A", "B. Vitamin B12", "C. Vitamin C", "D. Vitamin D"],
            "answer": "D",
            "explanation": "Kroppen kan produsere vitamin D når huden utsettes for sollys, noe som er viktig for beinhelsen."
        },
        {
            "question": "Hvilket mineral er viktig for nervefunksjon og muskelsammentrekning?",
            "options": ["A. Kalsium", "B. Jern", "C. Magnesium", "D. Sink"],
            "answer": "C",
            "explanation": "Magnesium er viktig for nervefunksjon og muskelsammentrekning, samt for energiproduksjon."
        },
        {
            "question": "Hvilket vitamin er viktig for hudens helse og sårheling?",
            "options": ["A. Vitamin A", "B. Vitamin B12", "C. Vitamin C", "D. Vitamin D"],
            "answer": "C",
            "explanation": "Vitamin C er viktig for hudens helse og sårheling, da det bidrar til dannelse av kollagen."
        },
        {
            "question": "Hvilke av følgende vitaminer er fettløslige?",
            "options": ["A. Vitamin A", "B. Vitamin B", "C. Vitamin C", "D. Vitamin D"],
            "answer": "D",
            "explanation": "Vitamin D er fettløselig og viktig for opptak av kalsium."
        }

]


# Initialiser GUI
root = tk.Tk()
root.title("Quiz App")
root.geometry("600x550")
style = Style(theme="flatly")
style.configure("TLabel", font=("Helvetica", 16))
style.configure("TButton", font=("Helvetica", 14))

# Globale variabler
score = 0
current_question = 0
choice_btns = []
quiz_data = []
user_answers = []

# Widgets
qs_label = ttk.Label(root, anchor="center", wraplength=500, padding=10)
qs_label.pack(pady=10)

choices_frame = ttk.Frame(root)
choices_frame.pack(pady=10, fill=tk.X)

feedback_label = ttk.Label(root, anchor="center", padding=10)
feedback_label.pack(pady=10)

score_label = ttk.Label(root, text="", anchor="center", padding=10)
score_label.pack(pady=10)

nav_frame = ttk.Frame(root)
nav_frame.pack(pady=10)

prev_btn = ttk.Button(nav_frame, text="Tilbake", command=lambda: previous_question(), state="disabled")
prev_btn.pack(side="left", padx=10)

next_btn = ttk.Button(nav_frame, text="Neste", command=lambda: next_question(), state="disabled")
next_btn.pack(side="left", padx=10)

retry_btn = ttk.Button(root, text="Prøv igjen", command=lambda: restart_quiz(), state="disabled")
retry_btn.pack(pady=10)

progress = ttk.Progressbar(root, orient="horizontal", length=400, mode="determinate")
progress.pack(pady=10)

# --- Funksjoner ---

def restart_quiz():
    global score, current_question, choice_btns, quiz_data, user_answers
    score = 0
    current_question = 0
    quiz_data = original_quiz_data.copy()
    random.shuffle(quiz_data)
    user_answers = [None] * len(quiz_data)
    progress.config(maximum=len(quiz_data), value=0)
    retry_btn.config(state="disabled")
    next_btn.config(state="disabled")
    prev_btn.config(state="disabled")
    show_question()

def show_question():
    global choice_btns
    for btn in choice_btns:
        btn.destroy()
    choice_btns = []

    question = quiz_data[current_question]
    qs_label.config(text=question["question"])
    options = question["options"]

    selected_index = user_answers[current_question]

    for i, option in enumerate(options):
        button = ttk.Button(
            choices_frame,
            text=option,
            command=lambda i=i: check_answer(i),
            style="TButton"
        )
        button.pack(pady=5, fill=tk.X)
        choice_btns.append(button)

    feedback_label.config(text="")
    score_label.config(text="Score: {}/{}".format(score, len(quiz_data)))
    prev_btn.config(state="normal" if current_question > 0 else "disabled")
    next_btn.config(state="disabled")
    retry_btn.config(state="disabled")

    if selected_index is not None:
        apply_feedback_colors(selected_index)
        next_btn.config(state="normal")

def check_answer(choice_index):
    global score
    question = quiz_data[current_question]
    selected_letter = chr(65 + choice_index)
    correct_letter = question["answer"]
    correct_index = ord(correct_letter) - 65
    previous = user_answers[current_question]

    # Juster score hvis bytter svar
    if previous is not None and chr(65 + previous) == correct_letter:
        score -= 1

    user_answers[current_question] = choice_index

    if selected_letter == correct_letter:
        score += 1
        feedback_label.config(text="Riktig! " + question['explanation'], foreground="green")
    else:
        feedback_label.config(
            text=f"Galt! Det riktige svaret er {correct_letter}. {question['explanation']}",
            foreground="red"
        )

    score_label.config(text="Score: {}/{}".format(score, len(quiz_data)))
    progress['value'] = current_question + 1

    apply_feedback_colors(choice_index)
    next_btn.config(state="normal")

def apply_feedback_colors(selected_index):
    question = quiz_data[current_question]
    correct_index = ord(question["answer"]) - 65

    for i, button in enumerate(choice_btns):
        button.config(state="disabled")
        if i == correct_index:
            button.config(style="success.TButton")
        elif i == selected_index:
            button.config(style="danger.TButton")
        else:
            button.config(style="secondary.TButton")

def next_question():
    global current_question
    if current_question < len(quiz_data) - 1:
        current_question += 1
        show_question()
    else:
        messagebox.showinfo(
            "Quiz gjennomført",
            f"Du fikk {score} av {len(quiz_data)} riktige.\nTakk for at du deltok!"
        )
        retry_btn.config(state="normal")
        next_btn.config(state="disabled")
        prev_btn.config(state="disabled")

def previous_question():
    global current_question
    if current_question > 0:
        current_question -= 1
        show_question()

# Start første gang
restart_quiz()
root.mainloop()
```
