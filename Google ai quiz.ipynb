```python
import locale
import tkinter as tk
from tkinter import messagebox, ttk
import random
from ttkbootstrap import Style

# Lokalitet-håndtering (som i originalkoden)
try:
    locale.setlocale(locale.LC_ALL, 'en_US.UTF-8')
except locale.Error:
    try:
        locale.setlocale(locale.LC_ALL, 'nb_NO.UTF-8')
    except locale.Error:
        locale.setlocale(locale.LC_ALL, '')

# --- DATA ---
# Det er god praksis å holde data adskilt fra applikasjonslogikken.
# For et større prosjekt kunne dette vært i en egen fil, f.eks. "quiz_data.py".
original_quiz_data = [
    # ... (all din quiz-data her, ingen endring nødvendig) ...
    {
        "question": "Hva er Alfred Wegener kjent for?",
        "options": [
            "A. Teorien om platedrift",
            "B. Teorien om havbunnspredning",
            "C. Teorien om kontinentaldrift",
            "D. Teorien om jordrotasjon"
        ],
        "answer": "C",
        "explanation": "Wegener lanserte teorien om kontinentaldrift i 1912, som senere ble utviklet teorien om platetektonikk."
    },
    {
        "question": "Hva kjennetegner en aktiv vulkan?",
        "options": [
            "A. Et kjegleformet fjell spyr ut røyk og aske",
            "B. Tyntflytende lava flyter ut fra et rundt krater",
            "C. Magma kommer ut på jordoverflata",
            "D. Bare gamle utbrudd registrert"
        ],
        "answer": "C",
        "explanation": "En aktiv vulkan kjennetegnes ved at magma strømmer ut til jordoverflaten."
    },
    {
        "question": "Hvordan dannes avsetningsbergarter?",
        "options": [
            "A. Der løsmasser kittes sammen til fast stein",
            "B. Der det er høytrykk",
            "C. Bare i havet",
            "D. Ved smelting og størkning"
        ],
        "answer": "A",
        "explanation": "Avsetningsbergarter oppstår når løsmasser presses sammen og herdes til fast stein over tid."
    },
    {
        "question": "Hva er sant om størkningsbergarter som granitt?",
        "options": [
            "A. Dannes på overflaten",
            "B. Kan være grov- eller finkornet",
            "C. Består alltid av glimmer og sand",
            "D. Finnes kun i fjellkjeder"
        ],
        "answer": "B",
        "explanation": "Størkningsbergarter varierer i kornstørrelse basert på hvor dypt magmaen størknet."
    },
    {
        "question": "Hva er en størkningsbergart?",
        "options": [
            "A. En bergart som dannes når magma størkner",
            "B. En bergart som dannes av trykk og varme",
            "C. En bergart som dannes av løsmasser",
            "D. En bergart som bare finnes på havbunnen"
        ],
        "answer": "A",
        "explanation": "Størkningsbergarter dannes når smeltet stein (magma) avkjøles og størkner."
    },
    {
        "question": "Hva er en 'gang' i geologisk forstand?",
        "options": [
            "A. En tunnel laget av mennesker",
            "B. En sprekk i jordskorpa der magma har størknet",
            "C. En elv av lava på jordoverflata",
            "D. Et område med myk berggrunn"
        ],
        "answer": "B",
        "explanation": "En gang er en sprekk der magma har presset seg inn og størknet."
    },
    {
        "question": "Hva kjennetegner en dypbergart?",
        "options": [
            "A. Den dannes på jordoverflata og har små krystaller",
            "B. Den dannes av løsmasser og er lagdelt",
            "C. Den størkner dypt nede og er ofte grovkornet",
            "D. Den består bare av ett mineral"
        ],
        "answer": "C",
        "explanation": "Dypbergarter dannes dypt i jordskorpa og avkjøles langsomt, noe som gir store mineralkorn."
    },
    {
        "question": "Hvorfor er dypbergarter ofte grovkornede?",
        "options": [
            "A. Fordi de inneholder store mengder vann",
            "B. Fordi de består av organisk materiale",
            "C. Fordi de avkjøles sakte og mineralene får tid til å vokse",
            "D. Fordi de utsettes for høy temperatur på overflata"
        ],
        "answer": "C",
        "explanation": "Langsom avkjøling gjør at mineralene vokser seg store og synlige – dette gir en grovkornet struktur."
    },
    {
        "question": "Hvilken av følgende er en vanlig dypbergart som kan dannes i dype ganger?",
        "options": [
            "A. Basalt",
            "B. Granitt",
            "C. Kalkstein",
            "D. Marmor"
        ],
        "answer": "B",
        "explanation": "Granitt er en vanlig dypbergart som dannes ved langsom avkjøling av magma i dype soner."
    },
    {
        "question": "Hvilke bergarter er vanlige gangbergarter?",
        "options": [
            "A. Diabas og doleritt",
            "B. Granitt og marmor",
            "C. Sandstein og skifer",
            "D. Basalt og kalkstein"
        ],
        "answer": "A",
        "explanation": "Diabas og doleritt er vanlige gangbergarter som størkner i ganger i jordskorpa."
    },
    {
        "question": "Hva menes med at en størkningsbergart er dannet i en 'dyp gang'?",
        "options": [
            "A. At den er dannet i en vulkan",
            "B. At magmaen størknet på jordoverflaten",
            "C. At magma størknet i en sprekk dypt nede i jordskorpa",
            "D. At den ble dannet av avsatt leire"
        ],
        "answer": "C",
        "explanation": "En bergart dannet i dyp gang har størknet i en sprekk langt under jordoverflaten."
    },
    {
        "question": "Hva vet vi om bergarter?",
        "options": [
            "A. De består alltid av størknet lava",
            "B. De består av én type mineral",
            "C. De består av ett eller flere mineraler",
            "D. De dannes i havet"
        ],
        "answer": "C",
        "explanation": "Bergarter er bygd opp av ett eller flere mineraler, og finnes i ulike geologiske sammenhenger."
    },
    {
        "question": "Hva består litosfæren av?",
        "options": [
            "A. Kontinentskorpe og havbunnskorpe",
            "B. Jordskorpa og øvre del av mantelen",
            "C. Jordskorpa og mantelens flytende del",
            "D. Hele mantelen"
        ],
        "answer": "B",
        "explanation": "Litosfæren består av jordskorpa og den øvre, faste delen av mantelen."
    },
    {
        "question": "Hva oppstår alltid ved et jordskjelv?",
        "options": [
            "A. Sprekker i overflaten",
            "B. Bølgebevegelser i jordskorpa",
            "C. Tsunamier",
            "D. Vulkanutbrudd"
        ],
        "answer": "B",
        "explanation": "Jordskjelv fører til bølgebevegelser i jordskorpa, men ikke nødvendigvis sprekker eller tsunamier"
    },
    {
        "question": "Hva er en forkastning?",
        "options": [
            "A. Område der havbunn bøyes ned",
            "B. Resultat av forskyvning av to plater",
            "C. Raskt strømmende lava",
            "D. En vulkan i utbrudd"
        ],
        "answer": "B",
        "explanation": "En forkastning er en sprekk i jordskorpa der det har skjedd en forskyvning mellom to plater."
    },
    {
        "question": "Hvordan ble Himalaya dannet?",
        "options": [
            "A. Heving av asiatisk plata",
            "B. Gammel havbunn ble presset opp",
            "C. Kollisjon mellom to kontinentplater",
            "D. Vulkanutbrudd"
        ],
        "answer": "C",
        "explanation": "Himalaya ble dannet ved kollisjon mellom Den indiske og Den eurasiske plata."
    },
    {
        "question": "Hva skjer ved en midthavsrygg?",
        "options": [
            "A. En havbunnsplate møter en kontinentplate",
            "B. Det dannes ny havbunn",
            "C. Det oppstår tsunamier",
            "D. Det dannes fjellkjeder"
        ],
        "answer": "B",
        "explanation": "Ved midthavsrygger presses magma opp og danner ny havbunn."
    },
    {
        "question": "Når ble den kaledonske fjellkjeden dannet?",
        "options": [
            "A. Ca. 60 mill. år siden (tertiær)",
            "B. Ca. 400 mill. år siden (devon)",
            "C. Ca. 250 mill. år siden (perm)",
            "D. Ca. 700 mill. år siden (prekambrium)"
        ],
        "answer": "B",
        "explanation": "Den kaledonske fjellkjeden ble dannet i devon for rundt 400 millioner år siden."
    },
    {
        "question": "Hva heter laget under litosfæren?",
        "options": [
            "A. Ytre kjerne",
            "B. Termosfæren",
            "C. Astenosfæren",
            "D. Hydrosfæren"
        ],
        "answer": "C",
        "explanation": "Astenosfæren er et delvis smeltet lag under litosfæren."
    },
    {
        "question": "Hvilken størkningsbergart er mest grovkornet?",
        "options": [
            "A. Dagbergartene",
            "B. Gangbergartene",
            "C. Dypbergartene",
            "D. Lava"
        ],
        "answer": "C",
        "explanation": "Dypbergartene har størknet langsomt dypt nede og er derfor grovkornet."
    },
    {
        "question": "Hva er en tsunami?",
        "options": [
            "A. Lava som flyter raskt",
            "B. Japansk jordskjelv",
            "C. En flodbølge forårsaket av jordskjelv",
            "D. Undersjøisk vulkan"
        ],
        "answer": "C",
        "explanation": "Tsunamier er store bølger forårsaket av havbunnsjordskjelv."
    },
    {
        "question": "Hvilken landform dannes når to kontinentplater kolliderer?",
        "options": [
            "A. Fjellkjede",
            "B. Vulkansk øybue",
            "C. Midthavsrygg",
            "D. Graben"
        ],
        "answer": "A",
        "explanation": "Kollisjon mellom kontinentplater fører til dannelse av fjellkjeder."
    },
    {
        "question": "Hva er årsaken til jordskorpeplatenes bevegelse?",
        "options": [
            "A. De trekkes ned i mantelen",
            "B. Jordskjelv skyver dem",
            "C. Gravitasjon fra månen",
            "D. Solvind"
        ],
        "answer": "A",
        "explanation": "Plater beveger seg blant annet fordi kald havbunn trekkes ned i mantelen."
    },
    {
        "question": "Hva skjedde under jordskjelvet i Sumatra 2004?",
        "options": [
            "A. Stillehavsplata kolliderte med Den filippinske plata",
            "B. Indisk-australske plata kolliderte med Burmaplata",
            "C. Indisk-australske plata kolliderte med Afrikanske plata",
            "D. En vulkan eksploderte"
        ],
        "answer": "B",
        "explanation": "Jordskjelvet oppstod ved en kollisjon mellom Indisk-australske og Burmaplata."
    },
    {
        "question": "Hvor finner vi bergarter påvirket av den kaledonske fjellkjedefoldningen?",
        "options": [
            "A. I Alpeområdet i Mellom-Europa",
            "B. I Himalaya i Asia",
            "C. På begge sider av Atlanterhavet/Norskehavet",
            "D. I Andesfjellene"
        ],
        "answer": "C",
        "explanation": "Disse bergartene finnes både i Norge og på østkysten av Nord-Amerika."
    },
    {
        "question": "Hvilken utslippstype fra vulkanutbrudd kan nå lengst fra utslippsstedet?",
        "options": [
            "A. Lavastrøm",
            "B. Pyroklastisk strøm",
            "C. Gjørmestrøm",
            "D. Askenedfall"
        ],
        "answer": "C",
        "explanation": "Gjørmestrømmer kan følge elver og nå langt fra vulkanen."
    },
    {
        "question": "Hvilken av disse er en omdannet bergart?",
        "options": [
            "A. Marmor",
            "B. Sandstein",
            "C. Basalt",
            "D. Granitt"
        ],
        "answer": "A",
        "explanation": "Marmor er en omdannet (metamorf) bergart, opprinnelig kalkstein."
    },
    {
        "question": "Hvilken bergartstype tilhører granitt?",
        "options": [
            "A. Størkningsbergart",
            "B. Avsetningsbergart",
            "C. Omdannet bergart",
            "D. Vulkanbergart"
        ],
        "answer": "A",
        "explanation": "Granitt er en dypbergart og tilhører størkningsbergartene."
    },
    {
        "question": "Hvor er det størst jordskjelvaktivitet?",
        "options": [
            "A. Midt på jordskorpeplatene",
            "B. Nær plategrensene",
            "C. Nær ekvator",
            "D. Ved polene"
        ],
        "answer": "B",
        "explanation": "De fleste jordskjelv skjer nær plategrensene."
    },
    {
        "question": "Hva er en søylestrøm?",
        "options": [
            "A. Varmt materiale som stiger fra mantelen",
            "B. Kokende vann fra vulkan",
            "C. Høy lavafontene",
            "D. En strøm av aske og røyk"
        ],
        "answer": "A",
        "explanation": "En søylestrøm (hotspot) er oppstrømming av varmt materiale fra mantelen."
    },
    {
        "question": "Hvilken jordskorpeplate bygger opp Andesfjellene?",
        "options": [
            "A. Nazcaplata",
            "B. Den arktiske plata",
            "C. Stillehavsplata",
            "D. Cocosplata"
        ],
        "answer": "A",
        "explanation": "Nazcaplata presses under Sør-Amerikanske plata og skaper Andesfjellene."
    },
    {
        "question": "Hva er jordskjelv?",
        "options": [
            "A. Rystelser med store konsekvenser",
            "B. Rystelser over styrke 4,5",
            "C. Vibrasjoner i jordoverflata pga. energiutløsning",
            "D. Bare sterke nok til å rive bygninger"
        ],
        "answer": "C",
        "explanation": "Jordskjelv er vibrasjoner som oppstår når spenning i jordskorpa frigjøres."
    },
    {
        "question": "Hva kjennetegner alpine landformer?",
        "options": [
            "A. Fjelltopper over 2000 m og snø",
            "B. Bratte dalsider for alpinanlegg",
            "C. Botner, tinder og egger",
            "D. Breer og vidder"
        ],
        "answer": "C",
        "explanation": "Alpine landformer har spisse fjell og dype botner etter iserosjon."
    },
    {
        "question": "Hvordan lager isbreer skuringsstriper?",
        "options": [
            "A. Stein i isen sliper på underlaget",
            "B. Breslam sliper underlaget",
            "C. Smeltevann polerer",
            "D. Trykk fra istykkelse"
        ],
        "answer": "A",
        "explanation": "Fastfrosne steiner i breen virker som sandpapir og lager striper i fjellet."
    },
    {
        "question": "Hva skjedde med Danmark under siste istid?",
        "options": [
            "A. Danmark var helt nediset",
            "B. Danmark var isfritt",
            "C. Danmark var delvis nediset",
            "D. Danmark hadde bare isbreer i nord"
        ],
        "answer": "C",
        "explanation": "Danmark var delvis dekket av is under siste istid."
    },
    {
        "question": "Hva er riktig rekkefølge i et delta?",
        "options": [
            "A. Frontavsetning, toppavsetning og bunnavsetning",
            "B. Bunnavsetning, frontavsetning og toppavsetning",
            "C. Toppavsetning, frontavsetning og bunnavsetning",
            "D. Toppavsetning, bunnavsetning og frontavsetning"
        ],
        "answer": "B",
        "explanation": "Vannet avsetter først det groveste materialet (bunn), så front og til slutt topp."
    },
    {
        "question": "Hvor er erosjonen sterkest i en elvesving?",
        "options": [
            "A. Midt i elva",
            "B. I yttersvingen",
            "C. I innersvingen",
            "D. Ved munningen"
        ],
        "answer": "B",
        "explanation": "Erosjonen er sterkest i yttersvingen hvor vannhastigheten er størst."
    },
    {
        "question": "Hva er det norske begrepet for canyon?",
        "options": [
            "A. Gjel",
            "B. Juv",
            "C. Jettegryte",
            "D. Dypdal"
        ],
        "answer": "B",
        "explanation": "Et juv er en dyp, trang dal skapt av elveerosjon – samme som canyon."
    },
    {
        "question": "Hva betyr det at ei elv meandrerer?",
        "options": [
            "A. Transporterer finkornet materiale",
            "B. Gjør store svinger fra side til side",
            "C. Forgreiner seg utover",
            "D. Har konstant vannføring"
        ],
        "answer": "B",
        "explanation": "Meandrerende elver danner svingete løp over tid."
    },
    {
        "question": "Hva dannes når en elv avsetter materiale i hav eller innsjø?",
        "options": [
            "A. Elvevifte",
            "B. Delta",
            "C. Elveslette",
            "D. Flommark"
        ],
        "answer": "B",
        "explanation": "Når en elv møter stillestående vann, dannes et delta av avsatt materiale."
    },
    {
        "question": "Hva kalles en isbre som dekker et større fjellområde?",
        "options": [
            "A. Platåbre",
            "B. Botnbre",
            "C. Dalbre",
            "D. Innlandsis"
        ],
        "answer": "A",
        "explanation": "Platåbreer dekker store fjellområder, som Jostedalsbreen."
    },
    {
        "question": "Hva er mest karakteristisk for viddelandskapet?",
        "options": [
            "A. U-daler og fjordsjøer",
            "B. Mykt avrundede høyder",
            "C. Botner og egger",
            "D. Bratte fjellsider"
        ],
        "answer": "B",
        "explanation": "Viddelandskap har slake og avrundede høyder og åpent landskap."
    },
    {
        "question": "Hva menes med gamle landformer?",
        "options": [
            "A. Utformet av havet",
            "B. Utformet før paleogen",
            "C. Laget av isbreer",
            "D. Dannes etter siste istid"
        ],
        "answer": "B",
        "explanation": "Gamle landformer er dannet før den siste store geologiske perioden – paleogen."
    },
    {
        "question": "Hva er en botn?",
        "options": [
            "A. En dal-ende",
            "B. En dyp sjø",
            "C. En skålformet form laget av botnbre",
            "D. En fjelltopp"
        ],
        "answer": "C",
        "explanation": "En botn er en halvsirkelformet forsenkning i fjellet skapt av iserosjon."
    },
    {
        "question": "Hva betyr stedsnavn med -øyr eller -aur (f.eks. Lærdalsøyri)?",
        "options": [
            "A. Dalklype",
            "B. Fjord",
            "C. Delta",
            "D. Elveslette"
        ],
        "answer": "C",
        "explanation": "Disse navnene stammer fra deltaområder der elver møter hav eller innsjø."
    },
    {
        "question": "Hva kalles store steiner som har blitt fraktet med isen?",
        "options": [
            "A. Kampestein",
            "B. Bunnmorene",
            "C. Flyttblokk",
            "D. Skuringsstein"
        ],
        "answer": "C",
        "explanation": "Flyttblokker er store steiner fraktet av is og etterlatt da isen smeltet."
    },
    {
        "question": "Hva mener vi med marin grense?",
        "options": [
            "A. Høyeste nivå havet hadde etter istiden",
            "B. Laveste saltinnhold i marin leire",
            "C. Hvor langt havet trengte inn",
            "D. Grense mellom ferskvann og saltvann"
        ],
        "answer": "A",
        "explanation": "Marin grense er det høyeste nivået havet nådde etter siste istid."
    },
    {
        "question": "Hvordan dannes ravinelandskap?",
        "options": [
            "A. Stopp i isens tilbakesmelting",
            "B. Bekker og elver eroderer leire",
            "C. Kvikkleireskred over lang tid",
            "D. Frostsprengning og smeltevann"
        ],
        "answer": "B",
        "explanation": "Raviner dannes når vann graver i leiravsetninger etter istiden."
    },
    {
        "question": "Hva beskriver dannelsen av en fjord?",
        "options": [
            "A. Norsk fenomen – derfor internasjonalt begrep",
            "B. En druknet U-dal",
            "C. Demmet opp av isranddelta",
            "D. En smeltevannsrenne"
        ],
        "answer": "B",
        "explanation": "Fjorder er U-daler som ble fylt med havvann etter isens tilbaketrekking."
    },
    {
        "question": "Hva er erosjon?",
        "options": [
            "A. Avgnaging, nedtæring av landskap",
            "B. Forvitring på stedet",
            "C. Oppløsning av mineraler",
            "D. Sprekker i berggrunnen"
        ],
        "answer": "A",
        "explanation": "Erosjon er prosessen hvor ytre krefter sliter ned jordoverflaten."
    }
]

# --- Applikasjonsklasse ---
class QuizApp:
    def __init__(self, root):
        self.root = root
        self.root.title("Quiz App")
        self.root.geometry("600x550")
        self.style = Style(theme="flatly")
        self.style.configure("TLabel", font=("Helvetica", 16))
        self.style.configure("TButton", font=("Helvetica", 14))

        # Attributter for å holde styr på quiz-tilstanden
        self.score = 0
        self.current_question_index = 0
        self.quiz_data = []
        self.user_answers = [] # Lagrer indeksen til brukerens svar for hvert spørsmål

        # Opprett widgets (GUI-elementer)
        self._create_widgets()

        # Start quizen
        self.restart_quiz()

    def _create_widgets(self):
        """Oppretter alle GUI-elementene én gang."""
        # Spørsmål-label
        self.qs_label = ttk.Label(self.root, anchor="center", wraplength=500, padding=10)
        self.qs_label.pack(pady=10)

        # Ramme for svarknapper
        self.choices_frame = ttk.Frame(self.root)
        self.choices_frame.pack(pady=10, fill=tk.X, padx=20)

        # OPTIMALISERING: Lag knappene én gang og gjenbruk dem
        self.choice_btns = []
        for i in range(4): # Antar maks 4 svaralternativer
            button = ttk.Button(self.choices_frame, command=lambda i=i: self._check_answer(i))
            button.pack(pady=5, fill=tk.X)
            self.choice_btns.append(button)

        # Tilbakemelding-label
        self.feedback_label = ttk.Label(self.root, anchor="center", padding=10, wraplength=500, justify="left")
        self.feedback_label.pack(pady=10)

        # Poeng-label
        self.score_label = ttk.Label(self.root, text="", anchor="center", padding=10)
        self.score_label.pack(pady=10)

        # Navigasjonsramme
        nav_frame = ttk.Frame(self.root)
        nav_frame.pack(pady=10)

        self.prev_btn = ttk.Button(nav_frame, text="Tilbake", command=self._previous_question)
        self.prev_btn.pack(side="left", padx=10)

        self.next_btn = ttk.Button(nav_frame, text="Neste", command=self._next_question)
        self.next_btn.pack(side="left", padx=10)

        # Prøv igjen-knapp
        self.retry_btn = ttk.Button(self.root, text="Prøv igjen", command=self.restart_quiz)
        self.retry_btn.pack(pady=10)

        # Fremdriftsbar
        self.progress = ttk.Progressbar(self.root, orient="horizontal", length=400, mode="determinate")
        self.progress.pack(pady=20)

    def restart_quiz(self):
        """Nullstiller og starter quizen på nytt."""
        self.score = 0
        self.current_question_index = 0
        self.quiz_data = random.sample(original_quiz_data, len(original_quiz_data)) # Stokker om
        self.user_answers = [None] * len(self.quiz_data)

        self.progress.config(maximum=len(self.quiz_data), value=0)
        self._show_question()

    def _show_question(self):
        """Viser det nåværende spørsmålet og oppdaterer GUI."""
        question_data = self.quiz_data[self.current_question_index]
        
        # Hent brukerens tidligere svar for dette spørsmålet, hvis det finnes
        user_choice = self.user_answers[self.current_question_index]

        # Oppdater GUI-elementer
        self.qs_label.config(text=question_data["question"])
        
        for i, button in enumerate(self.choice_btns):
            button.config(text=question_data["options"][i], state="normal", style="TButton")

        if user_choice is not None:
            # Hvis brukeren allerede har svart, vis fargene og lås knappene
            self._apply_feedback_colors(user_choice)
            self._update_feedback_label(user_choice)
        else:
            # Hvis spørsmålet er nytt, nullstill tilbakemeldingen
            self.feedback_label.config(text="")

        self._update_ui_state()

    def _check_answer(self, choice_index):
        """Sjekker svaret, oppdaterer poengsum og GUI."""
        # Hvis brukeren har svart på dette spørsmålet før, må vi kanskje justere poengsummen
        previous_answer_index = self.user_answers[self.current_question_index]
        if previous_answer_index is not None:
            # Nullstill poengsum for forrige svar på dette spørsmålet
            prev_question_data = self.quiz_data[self.current_question_index]
            correct_letter_prev = prev_question_data["answer"]
            if chr(65 + previous_answer_index) == correct_letter_prev:
                self.score -= 1
        
        # Lagre det nye svaret
        self.user_answers[self.current_question_index] = choice_index

        question_data = self.quiz_data[self.current_question_index]
        selected_letter = chr(65 + choice_index)
        correct_letter = question_data["answer"]

        # Gi poeng hvis riktig
        if selected_letter == correct_letter:
            self.score += 1
            
        # Vis farger og tilbakemelding
        self._apply_feedback_colors(choice_index)
        self._update_feedback_label(choice_index)
        self._update_ui_state()

    def _apply_feedback_colors(self, selected_index):
        """Oppdaterer fargen på knappene for å vise riktig/galt svar."""
        question_data = self.quiz_data[self.current_question_index]
        correct_index = ord(question_data["answer"]) - 65

        for i, button in enumerate(self.choice_btns):
            button.config(state="disabled") # Deaktiver alle knapper etter svar
            if i == correct_index:
                button.config(style="success.TButton")
            elif i == selected_index:
                button.config(style="danger.TButton")
            else:
                button.config(style="secondary.TButton")
    
    def _update_feedback_label(self, selected_index):
        """Oppdaterer teksten i feedback-labelen."""
        question_data = self.quiz_data[self.current_question_index]
        selected_letter = chr(65 + selected_index)
        correct_letter = question_data["answer"]

        if selected_letter == correct_letter:
            self.feedback_label.config(text=f"Riktig!\n{question_data['explanation']}", foreground="green")
        else:
            self.feedback_label.config(text=f"Galt! Riktig svar var {correct_letter}.\n{question_data['explanation']}", foreground="red")


    def _update_ui_state(self):
        """Sentral funksjon for å oppdatere tilstanden til alle dynamiske elementer."""
        # Oppdater poengsum og fremdrift
        self.score_label.config(text=f"Score: {self.score}/{len(self.quiz_data)}")
        answered_count = len([ans for ans in self.user_answers if ans is not None])
        self.progress.config(value=answered_count)

        # Sett tilstand på navigasjonsknapper
        self.prev_btn.config(state="normal" if self.current_question_index > 0 else "disabled")
        self.next_btn.config(state="normal" if self.current_question_index < len(self.quiz_data) - 1 else "disabled")

        # Aktiver "Prøv igjen" kun på slutten
        is_quiz_finished = self.current_question_index == len(self.quiz_data) - 1
        self.retry_btn.config(state="normal" if is_quiz_finished else "disabled")

        # Neste-knapp skal kun være aktiv hvis brukeren har svart
        if self.user_answers[self.current_question_index] is None:
            self.next_btn.config(state="disabled")


    def _next_question(self):
        """Går til neste spørsmål."""
        if self.current_question_index < len(self.quiz_data) - 1:
            self.current_question_index += 1
            self._show_question()
        else:
            self._show_final_score()

    def _previous_question(self):
        """Går til forrige spørsmål."""
        if self.current_question_index > 0:
            self.current_question_index -= 1
            self._show_question()

    def _show_final_score(self):
        """Viser en meldingsboks med sluttresultatet."""
        messagebox.showinfo(
            "Quiz gjennomført",
            f"Du fikk {self.score} av {len(self.quiz_data)} riktige.\nTakk for at du deltok!"
        )
        self.retry_btn.config(state="normal")
        self.next_btn.config(state="disabled")
        self.prev_btn.config(state="disabled")


if __name__ == "__main__":
    root = tk.Tk()
    app = QuizApp(root)
    root.mainloop()
```
