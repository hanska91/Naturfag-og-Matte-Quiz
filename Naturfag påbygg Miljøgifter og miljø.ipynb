```python
import locale
import tkinter as tk
from tkinter import messagebox, ttk
import random

# Lokalitet (for å unngå feil med ttkbootstrap)
try:
    locale.setlocale(locale.LC_ALL, 'en_US.UTF-8')
except locale.Error:
    locale.setlocale(locale.LC_ALL, '')

from ttkbootstrap import Style
locale.setlocale(locale.LC_ALL, '')

# Opprinnelige quiz-data (lagres separat for reset)
original_quiz_data = [
  {
    "question": "Hva er miljøgifter?",
    "options": ["A. Naturlige stoffer som er harmløse", "B. Stoffer som brytes raskt ned i naturen", "C. Kjemikalier som hoper seg opp i næringskjeder og er giftige over tid", "D. Utslipp fra planter"],
    "answer": "C",
    "explanation": "Miljøgifter er giftige stoffer som ikke brytes lett ned, og som kan lagres i levende organismer."
  },
  {
    "question": "Hva slags skade kan du få hvis du eksponeres for løsemidler over tid?",
    "options": ["A. Benskjørhet", "B. Nerveskader", "C. Hudkreft", "D. Økt appetitt"],
    "answer": "B",
    "explanation": "Langvarig eksponering for løsemidler kan føre til nerveskader, som blant annet kan gi konsentrasjonsvansker og hukommelsessvikt."
  },
  {
    "question": "Hva betyr FSC-merket på et treprodukt?",
    "options": ["A. At produktet er økologisk", "B. At det er svanemerket", "C. At det er laget av tre fra bærekraftig skogbruk", "D. At det er produsert i Norge"],
    "answer": "C",
    "explanation": "FSC-merket viser at produktet kommer fra bærekraftig skogbruk, der hensyn er tatt til miljø, mennesker og økonomi."
  },
  {
    "question": "Hva er navnet på en gruppe miljøgifter som inkluderer bly, kadmium og kvikksølv?",
    "options": ["A. Tungmetaller", "B. Løsemidler", "C. Hormonhermere", "D. Radioaktive stoffer"],
    "answer": "A",
    "explanation": "Tungmetaller som bly, kadmium og kvikksølv er miljøgifter som kan hope seg opp i kroppen og føre til alvorlige helseproblemer."
  },
  {
    "question": "Hva betyr 'cocktaileffekt' når det gjelder miljøgifter?",
    "options": ["A. At miljøgifter smaker som fruktcocktail", "B. At flere miljøgifter kan virke sammen og gi sterkere effekt", "C. At miljøgifter bare virker i blanding med vann", "D. At miljøgifter blir mindre farlige når de blandes"],
    "answer": "B",
    "explanation": "Cocktaileffekten betyr at flere miljøgifter kan virke sammen og gi større skade enn hver enkelt ville gjort alene."
  },
  {
    "question": "Hvordan kan vi som forbrukere bidra til å unngå miljøgifter?",
    "options": ["A. Ved å kjøpe brukte produkter og unngå plast og kosmetikk med farlige stoffer", "B. Ved å kjøpe mest mulig nytt", "C. Ved å bruke mer plantevernmidler", "D. Ved å kaste alt i restavfallet"],
    "answer": "A",
    "explanation": "Som forbrukere kan vi bidra ved å velge produkter uten farlige kjemikalier, for eksempel svanemerkede eller miljøsertifiserte varer."
  },
  {
    "question": "Hvilket av disse er et av FNs bærekraftsmål?",
    "options": ["A. Mer oljeproduksjon", "B. Rent vann og gode sanitærforhold", "C. Flere biler", "D. Mer bruk av kull"],
    "answer": "B",
    "explanation": "Rent vann og gode sanitærforhold er ett av FNs 17 bærekraftsmål som skal sikre en bærekraftig utvikling globalt."
  },
  {
    "question": "Hvilke problemer er forbundet med global oppvarming?",
    "options": ["A. Kaldere vintre og mer snø", "B. Redusert havnivå", "C. Tørke, ekstremvær og issmelting", "D. Bedre luftkvalitet"],
    "answer": "C",
    "explanation": "Global oppvarming fører til klimaendringer som gir tørke, ekstremvær og smelting av isbreer og polare strøk."
  },
  {
    "question": "Hva er en fornybar energikilde?",
    "options": ["A. Olje", "B. Kull", "C. Solenergi", "D. Naturgass"],
    "answer": "C",
    "explanation": "Solenergi er en fornybar energikilde fordi den ikke blir brukt opp og fornyes naturlig av sollyset."
  },
  {
    "question": "Hvilket yrke kan bidra til en mer bærekraftig framtid?",
    "options": ["A. Kullgruvearbeider", "B. Miljørådgiver", "C. Oljeboringsingeniør", "D. Bilselger"],
    "answer": "B",
    "explanation": "Miljørådgivere jobber med løsninger for miljø og bærekraft, og er viktige for å løse klima- og miljøutfordringer."
  },
    {
        "question": "Hvilken gruppe organismer er mest utsatt for miljøgifter i næringskjeder?",
        "options": ["A. Planter", "B. Førstekonsumenter", "C. Rovdyr høyt i næringskjeden", "D. Nedbrytere"],
        "answer": "C",
        "explanation": "Rovdyr høyt i næringskjeden får i seg mest miljøgifter fordi stoffene akkumuleres i hvert ledd."
    },
    {
        "question": "Hva er en livsstilssykdom?",
        "options": ["A. En sykdom man blir født med", "B. En sykdom som skyldes miljøgifter", "C. En sykdom knyttet til vaner som kosthold og fysisk aktivitet", "D. En infeksjonssykdom"],
        "answer": "C",
        "explanation": "Livsstilssykdommer skyldes ofte dårlig kosthold, inaktivitet og røyking – for eksempel diabetes type 2 og hjerte- og karsykdommer."
    }
]

# Initialiser GUI
root = tk.Tk()
root.title("Quiz App")
root.geometry("600x550")
style = Style(theme="flatly")
style.configure("TLabel", font=("Helvetica", 16))
style.configure("TButton", font=("Helvetica", 14))

# Globale variabler
score = 0
current_question = 0
choice_btns = []
quiz_data = []
user_answers = []

# Widgets
qs_label = ttk.Label(root, anchor="center", wraplength=500, padding=10)
qs_label.pack(pady=10)

choices_frame = ttk.Frame(root)
choices_frame.pack(pady=10, fill=tk.X)

feedback_label = ttk.Label(root, anchor="center", padding=10, wraplength=500, justify="left")
feedback_label.pack(pady=10)

score_label = ttk.Label(root, text="", anchor="center", padding=10)
score_label.pack(pady=10)

nav_frame = ttk.Frame(root)
nav_frame.pack(pady=10)

prev_btn = ttk.Button(nav_frame, text="Tilbake", command=lambda: previous_question(), state="disabled")
prev_btn.pack(side="left", padx=10)

next_btn = ttk.Button(nav_frame, text="Neste", command=lambda: next_question(), state="disabled")
next_btn.pack(side="left", padx=10)

retry_btn = ttk.Button(root, text="Prøv igjen", command=lambda: restart_quiz(), state="disabled")
retry_btn.pack(pady=10)

progress = ttk.Progressbar(root, orient="horizontal", length=400, mode="determinate")
progress.pack(pady=10)

# --- Funksjoner ---

def restart_quiz():
    global score, current_question, choice_btns, quiz_data, user_answers
    score = 0
    current_question = 0
    quiz_data = original_quiz_data.copy()
    random.shuffle(quiz_data)
    user_answers = [None] * len(quiz_data)
    progress.config(maximum=len(quiz_data), value=0)
    retry_btn.config(state="disabled")
    next_btn.config(state="disabled")
    prev_btn.config(state="disabled")
    show_question()

def show_question():
    global choice_btns
    for btn in choice_btns:
        btn.destroy()
    choice_btns = []

    question = quiz_data[current_question]
    qs_label.config(text=question["question"])
    options = question["options"]

    selected_index = user_answers[current_question]

    for i, option in enumerate(options):
        button = ttk.Button(
            choices_frame,
            text=option,
            command=lambda i=i: check_answer(i),
            style="TButton"
        )
        button.pack(pady=5, fill=tk.X)
        choice_btns.append(button)

    feedback_label.config(text="")
    score_label.config(text="Score: {}/{}".format(score, len(quiz_data)))
    prev_btn.config(state="normal" if current_question > 0 else "disabled")
    next_btn.config(state="disabled")
    retry_btn.config(state="disabled")

    if selected_index is not None:
        apply_feedback_colors(selected_index)
        next_btn.config(state="normal")

def check_answer(choice_index):
    global score
    question = quiz_data[current_question]
    selected_letter = chr(65 + choice_index)
    correct_letter = question["answer"]
    correct_index = ord(correct_letter) - 65
    previous = user_answers[current_question]

    # Juster score hvis bytter svar
    if previous is not None and chr(65 + previous) == correct_letter:
        score -= 1

    user_answers[current_question] = choice_index

    if selected_letter == correct_letter:
        score += 1
        feedback_label.config(text="Riktig!\n" + question['explanation'], foreground="green")
    else:
        feedback_label.config(
    text=f"Galt! Det riktige svaret er {correct_letter}.\n{question['explanation']}",
    foreground="red"
        )

    score_label.config(text="Score: {}/{}".format(score, len(quiz_data)))
    progress['value'] = current_question + 1

    apply_feedback_colors(choice_index)
    next_btn.config(state="normal")

def apply_feedback_colors(selected_index):
    question = quiz_data[current_question]
    correct_index = ord(question["answer"]) - 65

    for i, button in enumerate(choice_btns):
        button.config(state="disabled")
        if i == correct_index:
            button.config(style="success.TButton")
        elif i == selected_index:
            button.config(style="danger.TButton")
        else:
            button.config(style="secondary.TButton")

def next_question():
    global current_question
    if current_question < len(quiz_data) - 1:
        current_question += 1
        show_question()
    else:
        messagebox.showinfo(
            "Quiz gjennomført",
            f"Du fikk {score} av {len(quiz_data)} riktige.\nTakk for at du deltok!"
        )
        retry_btn.config(state="normal")
        next_btn.config(state="disabled")
        prev_btn.config(state="disabled")

def previous_question():
    global current_question
    if current_question > 0:
        current_question -= 1
        show_question()

# Start første gang
restart_quiz()
root.mainloop()
```
