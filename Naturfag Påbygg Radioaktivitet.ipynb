```python
import random
import tkinter as tk
from PIL import Image, ImageTk, UnidentifiedImageError
import requests
from io import BytesIO

def quiz():
    questions = [
        {
            "question": "Hvilket grunnstoff har 30 protoner i kjernen?",
            "options": ["A. Karbon (C)", "B. Hydrogen (H)", "C. Sink (Zn)", "D. Kobber (Cu)"],
            "answer": "C",
            "explanation": "Sink (Zn) har 30 protoner i kjernen, siden det har atomnummer 30 i periodesystemet.",
            "image": "https://upload.wikimedia.org/wikipedia/commons/0/0b/Zinc_crystals.jpg"
        },
        {
            "question": "Hva er halveringstid?",
            "options": ["A. Tiden det tar for et stoff å brytes ned", "B. Tiden det tar for stråling å forsvinne", "C. Tiden det tar for et atom å bli stabilt", "D. Tiden det tar før halvparten av atomkjernene har sendt ut radioaktiv stråling"],
            "answer": "D",
            "explanation": "Halveringstid er tiden det tar før halvparten av atomkjernene i et radioaktivt stoff har sendt ut radioaktiv stråling.",
            "image": "https://upload.wikimedia.org/wikipedia/commons/4/4e/Half_life_diagram.svg"
        },
        # Add more questions as needed...
    ]

    score = 0
    current_question_index = 0
    total_questions = len(questions)
    root = tk.Tk()
    root.title("Quiz")
    root.geometry("600x700")

    def show_question(index):
        nonlocal current_question_index
        if index < total_questions:
            current_question_index = index
            q = questions[index]
            question_label.config(text=q["question"])
            random.shuffle(q["options"])  # Randomize options order
            for i, option in enumerate(q["options"]):
                option_buttons[i].config(text=option, command=lambda opt=option: check_answer(opt, q))

            # Load the image
            if q.get("image"):
                try:
                    # Attempt to download and display the image
                    headers = {'User-Agent': 'Mozilla/5.0'}
                    response = requests.get(q["image"], headers=headers)
                    response.raise_for_status()  # Check if the URL is valid and accessible
                    image = Image.open(BytesIO(response.content))
                    image = image.resize((250, 250), Image.ANTIALIAS)  # Resize the image to fit the window
                    photo = ImageTk.PhotoImage(image)
                    image_label.config(image=photo)
                    image_label.image = photo  # Keep a reference to avoid garbage collection
                except (requests.RequestException, UnidentifiedImageError) as e:
                    print(f"Kunne ikke laste bildet: {e}")
                    image_label.config(text="Ingen bilde tilgjengelig")  # Display text if image is not available
            else:
                image_label.config(image='', text="Ingen bilde tilgjengelig")  # Display text if no image is provided

            # Start a countdown for answering the question
            countdown_timer(10)

    def countdown_timer(seconds):
        def update_timer():
            nonlocal seconds
            if seconds > 0:
                timer_label.config(text=f"Tid igjen: {seconds} sek.")
                seconds -= 1
                root.after(1000, update_timer)
            else:
                result_label.config(text="Tiden er ute!")
                explanation_label.config(text=questions[current_question_index]["explanation"])
                root.after(2000, lambda: show_next_question())  # Move to next question

        update_timer()

    def check_answer(selected_option, question):
        nonlocal score
        if selected_option[0] == question["answer"]:
            result_label.config(text="Riktig!", fg="green")
            score += 1
        else:
            result_label.config(text=f"Feil! Riktig svar er {question['answer']}.", fg="red")
        explanation_label.config(text=question["explanation"])
        
        # Show the next question immediately after answering
        root.after(2000, lambda: show_next_question())

    def show_next_question():
        nonlocal current_question_index
        if current_question_index + 1 < total_questions:
            show_question(current_question_index + 1)
        else:
            show_final_result()  # No more questions, show final score

    def show_final_result():
        final_score_label.config(text=f"Din endelige poengsum er {score} av {total_questions}.")
        restart_button.pack(pady=10)  # Show restart button

    def restart_quiz():
        nonlocal score, current_question_index
        score = 0
        current_question_index = 0
        restart_button.pack_forget()  # Hide restart button after restart
        show_question(0)  # Restart quiz

    # UI Elements
    question_label = tk.Label(root, text="", wraplength=500, font=('Arial', 14, 'bold'))
    question_label.pack(pady=20)

    option_buttons = [tk.Button(root, text="", wraplength=400, font=('Arial', 12)) for _ in range(4)]
    for button in option_buttons:
        button.pack(pady=5)

    image_label = tk.Label(root, font=('Arial', 12), anchor="center")
    image_label.pack(pady=20)

    timer_label = tk.Label(root, text="Tid igjen: 10 sek.", font=('Arial', 12))
    timer_label.pack(pady=10)

    result_label = tk.Label(root, text="", font=('Arial', 12))
    result_label.pack(pady=10)

    explanation_label = tk.Label(root, text="", wraplength=500, font=('Arial', 10))
    explanation_label.pack(pady=10)

    final_score_label = tk.Label(root, text="", font=('Arial', 14, 'bold'))
    final_score_label.pack(pady=20)

    restart_button = tk.Button(root, text="Start på nytt", command=restart_quiz, font=('Arial', 12))

    # Start quiz
    show_question(0)

    root.mainloop()

quiz()
```
