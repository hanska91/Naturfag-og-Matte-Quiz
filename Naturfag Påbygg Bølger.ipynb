```python
import locale
import tkinter as tk
from tkinter import messagebox, ttk
import random

# Lokalitet (for å unngå feil med ttkbootstrap)
try:
    locale.setlocale(locale.LC_ALL, 'en_US.UTF-8')
except locale.Error:
    locale.setlocale(locale.LC_ALL, '')

from ttkbootstrap import Style
locale.setlocale(locale.LC_ALL, '')

# Opprinnelige quiz-data (lagres separat for reset)
original_quiz_data = [
    {
        "question": "Hvilke to hovedtyper bølger har vi?",
        "options": ["A. Vannbølger og teoretiske bølger", "B. Statiske bølger og grafiske bølger", 
                    "C. Mekaniske bølger og elektromagnetiske bølger", "D. Solstråler og infrarød stråling"],
        "answer": "C",
        "explanation": "De to hovedtypene bølger er mekaniske bølger og elektromagnetiske bølger."
    },
    {
        "question": "Bølger frakter energi. Energien til en bølge er avhengig av...",
        "options": ["A. Solstråler", "B. Amplituden", "C. Bølgeretningen", "D. Temperaturen"],
        "answer": "B",
        "explanation": "Energien til en bølge er avhengig av amplituden – jo høyere amplitude, jo mer energi."
    },
    {
        "question": "Hvor fort beveger lyd seg i luft?",
        "options": ["A. 80 m/s", "B. 560 m/s", "C. 340 m/s", "D. 1200 m/s"],
        "answer": "C",
        "explanation": "Lyd beveger seg i luft med en fart på ca. 340 meter per sekund."
    },
    {
        "question": "Hva slags bølgefenomen bruker flaggermusa ved ekkolokalisering?",
        "options": ["A. Lydbølgene brytes", "B. Lydbølgene reflekteres", "C. Lydbølgene spres", "D. Lydbølgene fordamper"],
        "answer": "B",
        "explanation": "Flaggermus bruker refleksjon av lydbølger (ekko) for å finne bytte og orientere seg."
    },
    {
        "question": "Hvilke farger i det synlige lyset har kortest bølgelengde?",
        "options": ["A. Rød og oransje", "B. Gul og rød", "C. Blå og fiolett", "D. Grønn og blå"],
        "answer": "C",
        "explanation": "Blå og fiolett lys har kortest bølgelengde i det synlige spekteret."
    },
    {
        "question": "Hva er forskjellen på mekaniske og elektromagnetiske bølger?",
        "options": ["A. Bare elektromagnetiske bølger kan brytes", 
                    "B. Mekaniske bølger er mye raskere enn elektromagnetiske bølger", 
                    "C. Mekaniske bølger trenger et medium å spre seg i, mens elektromagnetiske bølger ikke trenger det", 
                    "D. De har forskjellig styrke på bølgene"],
        "answer": "C",
        "explanation": "Mekaniske bølger trenger et medium for å spre seg, mens elektromagnetiske bølger kan gå gjennom vakuum."
    },
    {
        "question": "Hva er eksempler på bølgefenomener?",
        "options": ["A. Spalting, refleksjon, resonans, bølgelengde, dopplereffekt, brytning, spredning", 
                    "B. Binding, spalting, brytning, temperatur", 
                    "C. Bølgehopp, sprett, absorpsjon, nøytralisering", 
                    "D. Lydnivå, bølgebevegelse, kraft, skygger"],
        "answer": "A",
        "explanation": "Typiske bølgefenomener inkluderer refleksjon, resonans, dopplereffekt, brytning og spredning."
    },
    {
        "question": "Hva slags type modulasjon er brukt her når bærebølgen endrer frekvens?",
        "options": ["A. Frekvensskiftmodulasjon", "B. Amplitudeskiftmodulasjon", 
                    "C. Frekvensmodulasjon", "D. Digital modulering"],
        "answer": "C",
        "explanation": "Når bærebølgen endrer frekvens, kalles det frekvensmodulasjon (FM)."
    },
    {
        "question": "Hva skjer med frekvensen til lydbølgene når vi snakker med lysere stemme?",
        "options": ["A. Frekvensen blir høyere", "B. Frekvensen blir lavere", 
                    "C. Frekvensen forsvinner", "D. Frekvensen påvirkes ikke"],
        "answer": "A",
        "explanation": "Lysere stemme betyr høyere frekvens på lydbølgene."
    },
    {
        "question": "Hvorfor oppstår dopplereffekt?",
        "options": ["A. Bølgene sprer seg i vakuum", 
                    "B. Lydbølgene sendes med infrarød stråling", 
                    "C. Den naturlige frekvensen treffes av energi med samme frekvens", 
                    "D. Kilden som sender ut bølgene beveger seg i forhold til oss"],
        "answer": "D",
        "explanation": "Dopplereffekten oppstår når bølgekilden beveger seg i forhold til observatøren."
    },
    {
        "question": "Hva er en bærebølge?",
        "options": ["A. En mottaker for radiobølger", 
                    "B. En EM-bølge med en bestemt frekvens som \"frakter\" det digitale signalet", 
                    "C. En radiobølge som ikke har frekvens", 
                    "D. En bølge som ikke kan reflekteres"],
        "answer": "B",
        "explanation": "En bærebølge er en elektromagnetisk bølge som brukes til å frakte informasjon via modulasjon."
    },
    {
        "question": "Bølgetyper som trenger et medium å spre seg i.",
        "answer": "mekaniske",
        "explanation": "Mekaniske bølger trenger partikler (et medium) for å kunne spre seg, i motsetning til elektromagnetiske bølger."
    },
    {
        "question": "Å endre bærebølgen for å kunne sende informasjon.",
        "answer": "modulere",
        "explanation": "Modulering betyr å endre en bærebølges frekvens eller amplitude for å sende informasjon."
    },
    {
        "question": "Bølgefenomenet som gjør at vi opplever at himmelen er blå.",
        "answer": "spredning",
        "explanation": "Lys spres i atmosfæren, og blått lys spres mer enn andre farger, derfor ser himmelen blå ut."
    },
    {
        "question": "Den synlige fargen med kortest bølgelengde.",
        "answer": "fiolett",
        "explanation": "Fiolett lys har den korteste bølgelengden i det synlige lysspekteret."
    },
    {
        "question": "Stråling som kan være nyttig for å kommunisere over korte avstander.",
        "answer": "infrarød",
        "explanation": "Infrarød stråling brukes i fjernkontroller og trådløs kommunikasjon over korte avstander."
    },
    {
        "question": "Måleenhet for frekvens.",
        "answer": "hertz",
        "explanation": "Frekvens måles i hertz (Hz), som er antall svingninger per sekund."
    },
    {
        "question": "Elektromagnetiske bølger kan spre seg i dette.",
        "answer": "vakuum",
        "explanation": "Elektromagnetiske bølger trenger ikke et medium – de kan spre seg i vakuum."
    },
    {
        "question": "Bølgefenomen som oppstår når kilden til bølgene flytter seg i forhold til oss.",
        "answer": "dopplereffekt",
        "explanation": "Dopplereffekt er endringen i frekvens når bølgekilden beveger seg i forhold til observatøren."
    },
    {
        "question": "Bølger frakter...",
        "answer": "energi",
        "explanation": "Alle bølger frakter energi fra ett sted til et annet."
    }
]

# Initialiser GUI
root = tk.Tk()
root.title("Quiz App")
root.geometry("600x550")
style = Style(theme="flatly")
style.configure("TLabel", font=("Helvetica", 16))
style.configure("TButton", font=("Helvetica", 14))

# Globale variabler
score = 0
current_question = 0
choice_btns = []
quiz_data = []
user_answers = []

# Widgets
qs_label = ttk.Label(root, anchor="center", wraplength=500, padding=10)
qs_label.pack(pady=10)

choices_frame = ttk.Frame(root)
choices_frame.pack(pady=10, fill=tk.X)

feedback_label = ttk.Label(root, anchor="center", padding=10, wraplength=500, justify="left")
feedback_label.pack(pady=10)

score_label = ttk.Label(root, text="", anchor="center", padding=10)
score_label.pack(pady=10)

nav_frame = ttk.Frame(root)
nav_frame.pack(pady=10)

prev_btn = ttk.Button(nav_frame, text="Tilbake", command=lambda: previous_question(), state="disabled")
prev_btn.pack(side="left", padx=10)

next_btn = ttk.Button(nav_frame, text="Neste", command=lambda: next_question(), state="disabled")
next_btn.pack(side="left", padx=10)

retry_btn = ttk.Button(root, text="Prøv igjen", command=lambda: restart_quiz(), state="disabled")
retry_btn.pack(pady=10)

progress = ttk.Progressbar(root, orient="horizontal", length=400, mode="determinate")
progress.pack(pady=10)

# --- Funksjoner ---

def restart_quiz():
    global score, current_question, choice_btns, quiz_data, user_answers
    score = 0
    current_question = 0
    quiz_data = original_quiz_data.copy()
    random.shuffle(quiz_data)
    user_answers = [None] * len(quiz_data)
    progress.config(maximum=len(quiz_data), value=0)
    retry_btn.config(state="disabled")
    next_btn.config(state="disabled")
    prev_btn.config(state="disabled")
    show_question()

def show_question():
    global choice_btns
    for btn in choice_btns:
        btn.destroy()
    choice_btns = []

    question = quiz_data[current_question]
    qs_label.config(text=question["question"])
    options = question["options"]

    selected_index = user_answers[current_question]

    for i, option in enumerate(options):
        button = ttk.Button(
            choices_frame,
            text=option,
            command=lambda i=i: check_answer(i),
            style="TButton"
        )
        button.pack(pady=5, fill=tk.X)
        choice_btns.append(button)

    feedback_label.config(text="")
    score_label.config(text="Score: {}/{}".format(score, len(quiz_data)))
    prev_btn.config(state="normal" if current_question > 0 else "disabled")
    next_btn.config(state="disabled")
    retry_btn.config(state="disabled")

    if selected_index is not None:
        apply_feedback_colors(selected_index)
        next_btn.config(state="normal")

def check_answer(choice_index):
    global score
    question = quiz_data[current_question]
    selected_letter = chr(65 + choice_index)
    correct_letter = question["answer"]
    correct_index = ord(correct_letter) - 65
    previous = user_answers[current_question]

    # Juster score hvis bytter svar
    if previous is not None and chr(65 + previous) == correct_letter:
        score -= 1

    user_answers[current_question] = choice_index

    if selected_letter == correct_letter:
        score += 1
        feedback_label.config(text="Riktig!\n" + question['explanation'], foreground="green")
    else:
        feedback_label.config(
    text=f"Galt! Det riktige svaret er {correct_letter}.\n{question['explanation']}",
    foreground="red"
        )

    score_label.config(text="Score: {}/{}".format(score, len(quiz_data)))
    progress['value'] = current_question + 1

    apply_feedback_colors(choice_index)
    next_btn.config(state="normal")

def apply_feedback_colors(selected_index):
    question = quiz_data[current_question]
    correct_index = ord(question["answer"]) - 65

    for i, button in enumerate(choice_btns):
        button.config(state="disabled")
        if i == correct_index:
            button.config(style="success.TButton")
        elif i == selected_index:
            button.config(style="danger.TButton")
        else:
            button.config(style="secondary.TButton")

def next_question():
    global current_question
    if current_question < len(quiz_data) - 1:
        current_question += 1
        show_question()
    else:
        messagebox.showinfo(
            "Quiz gjennomført",
            f"Du fikk {score} av {len(quiz_data)} riktige.\nTakk for at du deltok!"
        )
        retry_btn.config(state="normal")
        next_btn.config(state="disabled")
        prev_btn.config(state="disabled")

def previous_question():
    global current_question
    if current_question > 0:
        current_question -= 1
        show_question()

# Start første gang
restart_quiz()
root.mainloop()
```
