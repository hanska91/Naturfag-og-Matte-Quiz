```python
!pip install ttkbootstrap
import locale
import tkinter as tk
from tkinter import messagebox, ttk
import random

# Lokalitet (for å unngå feil med ttkbootstrap)
try:
    locale.setlocale(locale.LC_ALL, 'en_US.UTF-8')
except locale.Error:
    locale.setlocale(locale.LC_ALL, '')

from ttkbootstrap import Style
locale.setlocale(locale.LC_ALL, '')

# Opprinnelige quiz-data (lagres separat for reset)
original_quiz_data = [
    {
          "question": "Hva er en av hoveddimensjonene ved bærekraftig utvikling?",
          "options": ["A. Teknologisk utvikling", "B. Økonomisk vekst", "C. Sosial rettferdighet", "D. Politisk stabilitet"],
          "answer": "C",
          "explanation": "Bærekraftig utvikling handler om å møte dagens behov uten å ødelegge for fremtidige generasjoners evne til å møte sine egne behov. Dette innebærer å balansere tre hoveddimensjoner: økonomi, miljø og sosial rettferdighet."
    },
    {
          "question": "Hva innebærer bærekraftig utvikling?",
          "options": ["A. Å møte dagens behov uten å ødelegge for fremtidige generasjoners evne til å møte sine egne behov", "B. Å fokusere kun på økonomisk vekst", "C. Å prioritere teknologisk utvikling over miljøvern", "D. Å sikre politisk stabilitet"],
          "answer": "A",
          "explanation": "Bærekraftig utvikling krever altså en balanse mellom økonomisk vekst, miljøvern og sosial rettferdighet."
     },
     {
           "question": "Hvorfor er sosial rettferdighet viktig i bærekraftig utvikling?",
           "options": ["A. Fordi det sikrer at alle mennesker har tilgang til grunnleggende ressurser og muligheter", "B. Fordi det prioriterer økonomisk vekst over miljøvern", "C. Fordi det fokuserer på teknologisk utvikling", "D. Fordi det sikrer politisk stabilitet"],
           "answer": "A",
           "explanation": "Sosial rettferdighet er viktig i bærekraftig utvikling fordi det sikrer at alle mennesker har tilgang til grunnleggende ressurser og muligheter, og at ulikheter reduseres."
    },
    {
           "question": "Hva er en viktig faktor for å oppnå bærekraftig utvikling?",
           "options": ["A. Økt forbruk", "B. Redusert ressursbruk", "C. Økt industrialisering", "D. Redusert internasjonalt samarbeid"],
           "answer": "B",
           "explanation": "En viktig faktor for å oppnå bærekraftig utvikling er redusert ressursbruk. Dette innebærer å bruke ressurser på en måte som ikke utarmer dem for fremtidige generasjoner."
    },
    {
           "question": "Hvordan kan vi bidra til bærekraftig utvikling i hverdagen?",
           "options": ["A. Ved å øke vårt energiforbruk", "B. Ved å redusere avfall og resirkulere", "C. Ved å kjøpe flere engangsprodukter", "D. Ved å ignorere miljøspørsmål"],
           "answer": "B",
           "explanation": "Vi kan bidra til bærekraftig utvikling i hverdagen ved å redusere avfall og resirkulere. Dette hjelper med å bevare ressurser og redusere miljøpåvirkningen."
    },
    {
           "question": "Hva er en av hovedårsakene til klimaendringer?",
           "options": ["A. Naturlige svingninger i klimaet", "B. Menneskeskapte klimagassutslipp", "C. Redusert solaktivitet", "D. Økt vulkansk aktivitet"],
           "answer": "B",
           "explanation": "En av hovedårsakene til klimaendringer er menneskeskapte klimagassutslipp, som forsterker drivhuseffekten og fører til global oppvarming."
    },
    {
          "question": "Hvordan kan vi redusere vårt karbonavtrykk?",
          "options": ["A. Ved å øke vårt energiforbruk", "B. Ved å redusere avfall og resirkulere", "C. Ved å kjøpe flere engangsprodukter", "D. Ved å ignorere miljøspørsmål"],
          "answer": "B",
          "explanation": "Vi kan redusere vårt karbonavtrykk ved å redusere avfall og resirkulere, noe som hjelper med å bevare ressurser og redusere miljøpåvirkningen."
    },
    {
          "question": "Hva er en viktig faktor for bærekraftig ressursforvaltning?",
          "options": ["A. Økt ressursbruk", "B. Redusert ressursbruk", "C. Økt forurensning", "D. Redusert internasjonalt samarbeid"],
          "answer": "B",
          "explanation": "En viktig faktor for bærekraftig ressursforvaltning er redusert ressursbruk, som innebærer å bruke ressurser på en måte som ikke utarmer dem for fremtidige generasjoner."
    },
    {
          "question": "Hva er en av fordelene med globalisering?",
          "options": ["A. Økt økonomisk ulikhet", "B. Bedre tilgang til varer og tjenester", "C. Redusert internasjonalt samarbeid", "D. Økt miljøforurensning"],
          "answer": "B",
          "explanation": "En av fordelene med globalisering er bedre tilgang til varer og tjenester, noe som kan øke levestandarden."
    },
    {
          "question": "Hvordan kan globalisering påvirke bærekraftig utvikling?",
          "options": ["A. Ved å fremme økonomisk vekst uten hensyn til miljøet", "B. Ved å øke samarbeid og deling av bærekraftige teknologier", "C. Ved å redusere internasjonal handel", "D. Ved å øke nasjonal isolasjon"],
          "answer": "B",
          "explanation": "Globalisering kan påvirke bærekraftig utvikling ved å øke samarbeid og deling av bærekraftige teknologier og løsninger."
    },
    {
          "question": "Hva er demografisk overgang?",
          "options": ["A. En modell for økonomisk vekst", "B. En modell for befolkningsutvikling", "C. En modell for miljøvern", "D. En modell for politisk stabilitet"],
          "answer": "B",
          "explanation": "Demografisk overgang er en modell for hvordan befolkningen utvikler seg over tid, basert på faktorer som fødselsrate, dødsrate, innvandring og utvandring."
    },
    {
        "question": "Hva er en av hovedutfordringene knyttet til befolkningsvekst?",
        "options": ["A. Økt tilgang til ressurser", "B. Økt forurensning", "C. Redusert økonomisk vekst", "D. Økt politisk stabilitet"],
        "answer": "B",
        "explanation": "En av hovedutfordringene knyttet til befolkningsvekst er økt forurensning, som kan føre til miljøskader og helseproblemer."
    },
    {
        "question": "Hva er global oppvarming?",
        "options": ["A. En økning i jordas gjennomsnittstemperatur", "B. En reduksjon i jordas gjennomsnittstemperatur", "C. En konstant temperatur på jorda", "D. En tilfeldig variasjon i temperatur"],
        "answer": "A",
        "explanation": "Global oppvarming refererer til en økning i jordas gjennomsnittstemperatur, som har steget med 1,57 °C (2024) siden perioden 1880–1930."
    },
    {
        "question": "Hvor mye har temperaturen på Svalbard økt siden 1961?",
        "options": ["A. 1,11 °C", "B. 2 °C", "C. 5,6 °C", "D. 10 °C"],
        "answer": "C",
        "explanation": "Målinger på Svalbard viser at temperaturen har steget med 5,6 °C i gjennomsnitt siden 1961, noe som er nesten seks ganger mer enn den globale gjennomsnittstemperaturen."
    },
    {
        "question": "Hva kan global oppvarming føre til?",
        "options": ["A. Mindre ekstremvær", "B. Mer ekstremvær", "C. Konstant vær", "D. Ingen endring i værmønstre"],
        "answer": "B",
        "explanation": "Global oppvarming kan føre til mer ekstremvær, som store flommer, hetebølger og kraftige regnskyll."
    },
    {
        "question": "Hva er Milanković-sykluser?",
        "options": ["A. Regelmessige forandringer i jordas bane rundt sola og jordas rotasjonsakse", "B. Regelmessige forandringer i solens stråling", "C. Regelmessige forandringer i havstrømmer", "D. Regelmessige forandringer i atmosfærens sammensetning"],
        "answer": "A",
        "explanation": "Milanković-sykluser er regelmessige forandringer i jordas bane rundt sola og jordas rotasjonsakse, som påvirker mengden solinnstråling jorda mottar."
    },
    {
        "question": "Hva er en av hovedårsakene til den raske temperaturøkningen jorda nå opplever?",
        "options": ["A. Astronomiske endringer", "B. Platedrift", "C. Menneskeskapte klimagassutslipp", "D. Vulkanutbrudd"],
        "answer": "C",
        "explanation": "Den raske temperaturøkningen jorda nå opplever kan ikke forklares av naturlige faktorer alene, men skyldes i stor grad menneskeskapte klimagassutslipp."
    },
    {
        "question": "Hva er El Niño?",
        "options": ["A. En kald havstrøm langs kysten av Sør-Amerika", "B. En varm havstrøm som demper oppvellingen av kaldt havvann", "C. En periode med økt solinnstråling", "D. En periode med redusert vulkansk aktivitet"],
        "answer": "B",
        "explanation": "El Niño er en varm havstrøm som demper oppvellingen av kaldt havvann langs kysten av Sør-Amerika, noe som fører til høyere havtemperaturer og økt global lufttemperatur."
    },
        {
        "question": "Hva er hovedårsaken til den nåværende globale oppvarmingen?",
        "options": ["A. Økt solinnstråling", "B. Astronomiske endringer", "C. Vulkanutbrudd", "D. Utslipp av klimagasser"],
        "answer": "D",
        "explanation": "Den nåværende globale oppvarmingen skyldes hovedsakelig utslipp av klimagasser til atmosfæren etter den industrielle revolusjonen."
    },
    {
        "question": "Hvilken klimagass bidrar mest til global oppvarming?",
        "options": ["A. Vanndamp (H2O)", "B. Metan (CH4)", "C. Karbondioksid (CO2)", "D. Ozon (O3)"],
        "answer": "C",
        "explanation": "Karbondioksid (CO2) er den klimagassen som bidrar mest til global oppvarming."
    },
    {
        "question": "Hvor mye mer potent er metan i å fange varme i atmosfæren enn CO2 over en to-tiårsperiode?",
        "options": ["A) 28 ganger", "B) 34 ganger", "C) 84 ganger", "D) 165 ganger"],
        "answer": "C",
        "explanation": "Metan er 84 ganger mer potent enn CO2 i å fange varme i atmosfæren over en to-tiårsperiode."
    },
    {
        "question": "Hva er hovedkilden til menneskeskapte utslipp av lystgass?",
        "options": ["A) Fossilt brensel", "B) Avskoging", "C) Nitrogengjødsel og husdyrgjødsel", "D) Industrielle prosesser"],
        "answer": "C",
        "explanation": "Lystgassutslipp er hovedsakelig forbundet med bruk av nitrogengjødsel og husdyrgjødsel fra landbrukets ekspansjon og intensivering."
    },
    {
        "question": "Hvor mye har atmosfæriske konsentrasjoner av metan økt siden 1750?",
        "options": ["A) 110%", "B) 50%", "C) 80%","D) 165%"],
        "answer": "D",
        "explanation": "Atmosfæriske konsentrasjoner av metan er nå mer enn 165% høyere enn førindustrielle nivåer i 1750."
    },
    {
        "question": "Hvor mye har atmosfæriske konsentrasjoner av lystgass, eller dinitrogenoksid (N2O) økt siden 1750?",
        "options": ["A) 10%", "B) 15%", "C) 20%", "D) 25%"],
        "answer": "D",
        "explanation": "Atmosfæriske konsentrasjoner av lystgass, eller dinitrogenoksid (N2O) er nå 25% høyere enn førindustrielle nivåer i 1750. Den misbrukes ofte som ett rusmiddel"
    },
    {
        "question": "Hvor mye har mengden karbondioksid økt siden den industrielle revolusjonen?",
        "options": ["A. 33%", "B. 41%", "C. 51%", "D. 62%"],
        "answer": "C",
        "explanation": "Mengden karbondioksid i atmosfæren har økt med 51% siden den industrielle revolusjonen."
    },
    {
        "question": "Hva skjer når mengden klimagasser i atmosfæren øker?",
        "options": ["A. Mer varmestråling fra jorda absorberes", "B. Mindre varmestråling fra jorda absorberes", "C. Temperaturen synker", "D. Solinnstrålingen øker"],
        "answer": "A",
        "explanation": "Når mengden klimagasser i atmosfæren øker, absorberes mer av varmestrålingen fra jorda, noe som fører til økt temperatur."
    },
    {
        "question": "Hvor mye raskere endrer klimaet seg nå sammenlignet med slutten av siste istid?",
        "options": ["A. 5 ganger raskere", "B. 10 ganger raskere", "C. 15 ganger raskere", "D. 20 ganger raskere"],
        "answer": "D",
        "explanation": "Klimaet endrer seg nå opptil 20 ganger raskere enn oppvarmingen i slutten av siste istid."
    },
    {
        "question": "Hva er en av de viktigste forklaringene på temperaturstigningen over tid?",
        "options": ["A. Vulkanutbrudd", "B. Små endringer i solinnstråling", "C. Utslipp av klimagasser fra menneskelig aktivitet", "D. El Niño og La Niña"],
        "answer": "C",
        "explanation": "En av de viktigste forklaringene på temperaturstigningen over tid er utslipp av klimagasser fra menneskelig aktivitet."
    },
    {
        "question": "Hvordan påvirker lysere flater som snø og is solinnstrålingen?",
        "options": ["A. De absorberer mer solinnstråling", "B. De reflekterer mer solinnstråling", "C. De har ingen effekt på solinnstrålingen", "D. De øker solinnstrålingen"],
        "answer": "B",
        "explanation": "Lysere flater som snø og is reflekterer mer av solinnstrålingen tilbake."
    },
    {
        "question": "Hvor måles CO2-nivået som er på det høyeste på millioner av år?",
        "options": ["A. Mauna Loa på Hawaii", "B. Mount Everest", "C. Sahara-ørkenen", "D. Amazonas-regnskogen"],
        "answer": "A",
        "explanation": "CO2-nivået måles på Mauna Loa på Hawaii, og det er nå på det høyeste på millioner av år."
    },
    {
        "question": "Hvordan vet vi at temperaturen på jorda har økt?",
        "options": ["A. Gjennom observasjoner av dyreliv", "B. Gjennom målinger av lufttemperatur", "C. Gjennom endringer i vegetasjon", "D. Gjennom historiske dokumenter"],
        "answer": "B",
        "explanation": "Vi vet at temperaturen på jorda har økt gjennom målinger av lufttemperatur som startet i 1850."
    },
    {
        "question": "Hva brukes klimamodeller til?",
        "options": ["A. Å forutsi været for en enkelt dag", "B. Å beregne framtidens klima", "C. Å måle nåværende temperaturer", "D. Å overvåke vulkanutbrudd"],
        "answer": "B",
        "explanation": "Klimamodeller brukes til å beregne framtidens klima ved hjelp av matematiske ligninger."
    },
       {
        "question": "Hvor kommer den økende mengden av klimagasser i atmosfæren fra?",
        "options": ["A. Fotosyntese", "B. Karbonets kretsløp", "C. Fossile brensler", "D. Havvann"],
        "answer": "C",
        "explanation": "Den økende mengden av klimagasser i atmosfæren kommer hovedsakelig fra menneskelige aktiviteter som forbrenning av fossile brensler."
    },
    {
        "question": "Hva er karbonets kretsløp?",
        "options": ["A. En prosess der karbon lagres i havet", "B. En prosess der karbon flyttes mellom ulike lagre på jorda", "C. En prosess der karbon frigjøres fra fossile brensler", "D. En prosess der karbon lagres i atmosfæren"],
        "answer": "B",
        "explanation": "Karbonets kretsløp er en naturlig prosess der karbon flyttes mellom ulike lagre på jorda, inkludert atmosfæren, havet, og jordskorpa."
    },
    {
        "question": "Hvordan påvirker menneskelige aktiviteter karbonets kretsløp?",
        "options": ["A. Ved å øke fotosyntesen", "B. Ved å redusere mengden karbon i atmosfæren", "C. Ved å flytte karbon fra jordskorpa til atmosfæren", "D. Ved å øke karbonlagrene i havet"],
        "answer": "C",
        "explanation": "Menneskelige aktiviteter som forbrenning av fossile brensler flytter karbon fra jordskorpa til atmosfæren, noe som øker mengden klimagasser."
    },
    {
        "question": "Hva skjer når planter og dyr dør?",
        "options": ["A. De frigjør CO2 til atmosfæren eller havet", "B. De lagrer karbon i jordskorpa", "C. De øker mengden metan i atmosfæren", "D. De reduserer mengden karbon i økosystemene"],
        "answer": "A",
        "explanation": "Når planter og dyr dør, brytes de ned og frigjør CO2 til atmosfæren eller havet."
    },
    {
        "question": "Hva er en betydelig kilde til klimagassutslipp i tropiske strøk?",
        "options": ["A. Fotosyntese", "B. Avskoging", "C. Økt havtemperatur", "D. Økt plantevekst"],
        "answer": "B",
        "explanation": "Avskoging i tropiske strøk er en betydelig kilde til klimagassutslipp fordi det frigjør lagret karbon i trærne til atmosfæren."
    },
    {
        "question": "Hva er hovedkilden til CO2-utslipp på verdensbasis?",
        "options": ["A. Olje", "B. Kull", "C. Gass", "D. Avskoging"],
        "answer": "B",
        "explanation": "Bruk av kull i industrien og til elektrisitetsproduksjon er den største kilden til CO2-utslipp på verdensbasis."
    },
    {
        "question": "Hvor mye CO2 har blitt sluppet ut til atmosfæren siden 1750?",
        "options": ["A. 1000 gigatonn", "B. 1500 gigatonn", "C. 2000 gigatonn", "D. 2500 gigatonn"],
        "answer": "C",
        "explanation": "Siden 1750 har det blitt sluppet ut ca. 2000 gigatonn CO2 til atmosfæren."
    },
    {
        "question": "Hva skjer med CO2 i havvannet?",
        "options": ["A. Det lagres permanent", "B. Det frigjøres til atmosfæren", "C. Det brukes i fotosyntese", "D. Det omdannes til metan"],
        "answer": "C",
        "explanation": "CO2 i havvannet blir en del av næringskjedene når plantene bruker det i sin fotosyntese."
    },
    {
        "question": "Hva er en naturlig CO2-lager på den nordlige halvkula?",
        "options": ["A. Regnskogen", "B. Tundraen", "C. Havvannet", "D. Atmosfæren"],
        "answer": "B",
        "explanation": "Tundraen på den nordlige halvkula fungerer som et naturlig CO2-lager på grunn av det kalde klimaet som bremser nedbrytingen."
    },
    {
        "question": "Hva er en betydelig kilde til metanutslipp?",
        "options": ["A. Fotosyntese", "B. Husdyr og gjødsel", "C. Avskoging", "D. Oljeutvinning"],
        "answer": "B",
        "explanation": "Husdyr og gjødsel er en betydelig kilde til metanutslipp, spesielt fra fordøyelsesgasser fra kyr og sauer."
    },
    {
        "question": "Hva har ført til at vekstsesongen i Norge har blitt lengre?",
        "options": ["A. Økt nedbør", "B. Høyere temperaturer", "C. Mindre snø", "D. Mer sollys"],
        "answer": "B",
        "explanation": "Vekstsesongen i Norge har blitt lengre på grunn av mildere klima med høyere temperaturer."
    },
    {
        "question": "Hva er en konsekvens av at hvitkinngåsa har tilpasset seg klimaendringene?",
        "options": ["A. Den har blitt utryddet", "B. Den har flyttet sørover", "C. Den spiser opp avlingene på jordene", "D. Den har sluttet å migrere"],
        "answer": "C",
        "explanation": "Hvitkinngåsa har tilpasset seg klimaendringene, men har blitt en utfordring for bøndene fordi den spiser opp avlingene på jordene."
    },
    {
        "question": "Hvor mange flere mennesker vil være utsatt for risiko knyttet til havnivåstigning ved 2 °C enn ved 1,5 °C?",
        "options": ["A. 5 millioner", "B. 10 millioner", "C. 15 millioner", "D. 20 millioner"],
        "answer": "B",
        "explanation": "Ved 2 °C vil 10 millioner flere mennesker være utsatt for risiko knyttet til havnivåstigning enn ved 1,5 °C."
    },
    {
        "question": "Hva skjer når mengden klimagasser øker i atmosfæren?",
        "options": ["A. Drivhuseffekten forsterkes", "B. Temperaturen synker", "C. Nedbøren øker", "D. Havene blir surere"],
        "answer": "A",
        "explanation": "Når mengden klimagasser øker i atmosfæren, forsterkes drivhuseffekten, noe som gjør at det blir varmere på jorda."
    },
    {
        "question": "Hva er tilbakekobling i klimasystemet?",
        "options": ["A. En mekanisme som demper temperaturstigningen", "B. En mekanisme som forsterker temperaturstigningen", "C. En mekanisme som reduserer klimagassutslipp", "D. En mekanisme som øker nedbøren"],
        "answer": "B",
        "explanation": "Tilbakekobling er en mekanisme i klimasystemet som kan forsterke temperaturstigningen, for eksempel ved at økt vanndamp i atmosfæren fører til høyere temperaturer."
    },
    {
        "question": "Hva kan skje hvis temperaturen stiger med 2 °C?",
        "options": ["A. Alle korallrev vil forsvinne", "B. Isen i Antarktis vil smelte helt", "C. Havene vil bli surere", "D. Mange arter vil dø ut"],
        "answer": "D",
        "explanation": "En temperaturstigning på 2 °C kan føre til at mange arter, inkludert korallrev, dør ut."
    },
    {
        "question": "Hva er en viktig drivkraft for den nordatlantiske havstrømmen?",
        "options": ["A. Økt nedbør", "B. Dypvannsdannelse", "C. Mindre salt havvann", "D. Økt solinnstråling"],
        "answer": "B",
        "explanation": "Dypvannsdannelse, der kaldt og salt havvann synker mot havbunnen, er en viktig drivkraft for den nordatlantiske havstrømmen."
    },
    {
        "question": "Hva skjer når havvannet tar opp CO2 fra atmosfæren?",
        "options": ["A. Havvannet blir surere", "B. Havvannet blir varmere", "C. Havvannet blir kaldere", "D. Havvannet blir saltere"],
        "answer": "A",
        "explanation": "Når havvannet tar opp CO2 fra atmosfæren, blir det surere på grunn av dannelsen av karbonsyre."
    },
    {
        "question": "Hvor mye har havet steget siden 1880?",
        "options": ["A. 10 cm", "B. 20 cm", "C. 30 cm", "D. 40 cm"],
        "answer": "B",
        "explanation": "Havet har steget med mer enn 20 cm i gjennomsnitt siden 1880."
    },
    {
        "question": "Hva kan vi forvente oss i framtiden med hensyn til ekstremvær?",
        "options": ["A. Mindre nedbør", "B. Flere hetebølger", "C. Mindre tørke", "D. Stabilt klima"],
        "answer": "B",
        "explanation": "I framtiden kan vi forvente flere hetebølger og perioder med ekstremvær som følge av klimaendringene."
    },
    {
        "question": "Hva er en av hovedårsakene til at havnivået stiger?",
        "options": ["A. Økt nedbør", "B. Smelting av innlandsisene på Grønland og i Antarktis", "C. Økt plantevekst", "D. Mindre solinnstråling"],
        "answer": "B",
        "explanation": "Smelting av de store innlandsisene på Grønland og i Antarktis har stor betydning for havnivået."
    },
    {
        "question": "Hvor mye har havet steget i gjennomsnitt siden 1880?",
        "options": ["A. 10 cm", "B. 20 cm", "C. 30 cm", "D. 40 cm"],
        "answer": "B",
        "explanation": "Havet har steget med mer enn 20 cm i gjennomsnitt siden 1880."
    },
    {
        "question": "Hvor mye kan havnivået stige i dette århundret hvis klimagassutslippene fortsetter som i dag?",
        "options": ["A. 50 cm", "B. 60 cm", "C. 70 cm", "D. 80 cm"],
        "answer": "D",
        "explanation": "Hvis klimagassutslippene fortsetter som i dag, kan havnivået stige opp mot 80 cm i dette århundret."
    },
    {
        "question": "Hva er en konsekvens av at havvannet varmes opp?",
        "options": ["A. Havvannet blir surere", "B. Havvannet utvider seg og tar større plass", "C. Havvannet blir kaldere", "D. Havvannet blir saltere"],
        "answer": "B",
        "explanation": "Når havvannet varmes opp, utvider det seg og tar større plass, noe som bidrar til havnivåstigning."
    },
    {
        "question": "Hvor mange mennesker bor i lavtliggende kystområder som er utsatt ved et høyere havnivå?",
        "options": ["A. 500 millioner", "B. 1 milliard", "C. 1,5 milliarder", "D. 2 milliarder"],
        "answer": "B",
        "explanation": "Mer enn en milliard mennesker bor i lavtliggende kystområder som er utsatt ved et høyere havnivå."
    },
    {
        "question": "Hva kan være en løsning for å håndtere flom i norske kommuner?",
        "options": ["A. Bygge flere hus nær elva", "B. Øke nedbøren", "C. Bygge flomvern", "D. Redusere vegetasjonen"],
        "answer": "C",
        "explanation": "Å bygge flomvern mellom elveløpet og omgivelsene kan være en løsning for å håndtere flom i norske kommuner."
    },
    {
        "question": "Hva er en av de største truslene mot naturmangfoldet?",
        "options": ["A. Økt plantevekst", "B. Klimaendringer", "C. Mindre nedbør", "D. Økt solinnstråling"],
        "answer": "B",
        "explanation": "Klimaendringer er en av de største truslene mot naturmangfoldet, da det påvirker arter, økosystemer og naturtyper."
    },
    {
        "question": "Hva kan skje med arter som lever i områder med kaldt klima?",
        "options": ["A. De vil tilpasse seg raskt", "B. De vil få flere leveområder", "C. De vil dø ut eller bli færre", "D. De vil flytte sørover"],
        "answer": "C",
        "explanation": "Arter som lever i områder med kaldt klima kan dø ut eller bli færre hvis de ikke klarer å tilpasse seg klimaendringene."
    },
    {
        "question": "Hva kan skje med fiskebestander som følge av klimaendringer?",
        "options": ["A. De vil flytte seg nordover", "B. De vil øke i antall", "C. De vil flytte seg sørover", "D. De vil forbli uendret"],
        "answer": "A",
        "explanation": "Fiskebestander kan flytte seg nordover mot arktiske strøk som følge av klimaendringer."
    },
    {
        "question": "Hva kan være en konsekvens av økt havnivå for tett befolkede områder?",
        "options": ["A. Økt plantevekst", "B. Mindre nedbør", "C. Bosetting kan bli satt under vann", "D. Økt solinnstråling"],
        "answer": "C",
        "explanation": "Tett befolkede områder kan bli satt under vann som følge av økt havnivå."
    },
    {
        "question": "Hva har Trondheim gjort for å hindre oversvømmelser?",
        "options": ["A. Bygget flere hus", "B. Åpnet bekker og anlagt dammer og regnbed", "C. Økt asfaltdekningen", "D. Redusert vegetasjonen"],
        "answer": "B",
        "explanation": "Trondheim har åpnet bekker og anlagt dammer og regnbed for å holde igjen store vannmengder og hindre oversvømmelser."
    },
    {
        "question": "Hva kan grønne tak og vegger bidra til i byer?",
        "options": ["A. Øke nedbøren", "B. Kjøle ned bygningene og lagre nedbøren", "C. Redusere vegetasjonen", "D. Øke asfaltdekningen"],
        "answer": "B",
        "explanation": "Grønne tak og vegger kan både kjøle ned bygningene og lagre nedbøren som kommer."
    },
    {
        "question": "Hva er en del av klimatilpasningen i jordbruket?",
        "options": ["A. Økt bruk av asfalt", "B. Foredling av plantesorter som tåler endrede forhold", "C. Redusert nedbør", "D. Økt bruk av tette flater"],
        "answer": "B",
        "explanation": "Foredling av plantesorter som tåler endrede forhold er en del av klimatilpasningen i jordbruket."
    },
    {
        "question": "Hva kan grønne områder med bekker og små vann bidra til i boligområder?",
        "options": ["A. Øke asfaltdekningen", "B. Redusere nedbøren", "C. Gi en sosial møteplass og tilpasning til et endret klima", "D. Øke vegetasjonen"],
        "answer": "C",
        "explanation": "Grønne områder med bekker og små vann kan gi en sosial møteplass og tilpasning til et endret klima."
    },
    {
        "question": "Hva kan være en økonomisk gevinst ved klimatilpasningstiltak?",
        "options": ["A. Økte skader fra nedbør", "B. Redusert vegetasjon", "C. Unngå skader fra nedbør", "D. Økt asfaltdekning"],
        "answer": "C",
        "explanation": "En økonomisk gevinst ved klimatilpasningstiltak kan være å unngå skader fra nedbør."
    },
    {
        "question": "Hva kan være en del av klimatilpasningen i jordbruket?",
        "options": ["A. Økt bruk av asfalt", "B. Håndtering av nye skadedyr og plantesykdommer", "C. Redusert nedbør", "D. Økt bruk av tette flater"],
        "answer": "B",
        "explanation": "Håndtering av nye skadedyr og plantesykdommer må være en del av tilpasningen i jordbruket."
    },
    {
        "question": "Hva er hovedårsaken til dagens klimaendringer?",
        "options": ["A. Økt solinnstråling", "B. Drivhuseffekten har blitt sterkere", "C. Mindre vegetasjon", "D. Økt nedbør"],
        "answer": "B",
        "explanation": "Hovedårsaken til dagens klimaendringer er at drivhuseffekten har blitt sterkere på grunn av økt mengde drivhusgasser i atmosfæren."
    },
    {
        "question": "Hva er kilden til de ekstra drivhusgassene?",
        "options": ["A. Fornybare energikilder", "B. Fossile brensler", "C. Økt plantevekst", "D. Mindre nedbør"],
        "answer": "B",
        "explanation": "Kilden til de ekstra drivhusgassene er fossile brensler som olje, kull og gass."
    },
    {
        "question": "Hvor mye har havet steget med over tre millimeter i året?",
        "options": ["A. På grunn av økt nedbør", "B. På grunn av smelting av isbreer og ismassiver", "C. På grunn av økt solinnstråling", "D. På grunn av mindre vegetasjon"],
        "answer": "B",
        "explanation": "Havet stiger med over tre millimeter i året på grunn av smelting av isbreer og ismassiver på Grønland og i Antarktis."
    },
    {
        "question": "Hva skjer med ekstremregnet i Norge?",
        "options": ["A. Det blir mindre ekstremregn", "B. Det blir ingen endring", "C. Det blir mer ekstremregn", "D. Det blir tørrere"],
        "answer": "C",
        "explanation": "I Norge er det målt en sterk økning i antall dager med veldig mye regn, noe som fører til mer ekstremregn."
    },
    {
        "question": "Hva er en av de største utfordringene med klimaendringene?",
        "options": ["A. Å redusere nedbøren", "B. Å sikre gode liv og vilkår samtidig som utslippene går til null", "C. Å øke planteveksten", "D. Å redusere solinnstrålingen"],
        "answer": "B",
        "explanation": "En av de største utfordringene med klimaendringene er å sikre gode liv og vilkår samtidig som utslippene av drivhusgasser går til null."
    },
    {
        "question": "Hva må utslippene av drivhusgasser reduseres med hvert år fra i dag?",
        "options": ["A. 1-2 prosent", "B. 3-5 prosent", "C. 5-7 prosent", "D. 7-10 prosent"],
        "answer": "B",
        "explanation": "Utslippene av drivhusgasser må reduseres med minst 3-5 prosent i året fra i dag for å nå målet om å begrense oppvarmingen til to grader."
    },
    {
        "question": "Hva er en mulig konsekvens av at de frosne delene av Sibir tiner?",
        "options": ["A. Økt plantevekst", "B. Redusert nedbør", "C. Utslipp av metan", "D. Økt solinnstråling"],
        "answer": "C",
        "explanation": "Hvis de frosne delene av Sibir tiner, vil bakken råtne og slippe ut metan, som er en veldig kraftig drivhusgass."
    },
    {
        "question": "Hva må på plass i løpet av de neste 50 årene for å håndtere klimaendringene?",
        "options": ["A. Økt bruk av fossile brensler", "B. Fornybar energi, kraft fra elektrisitet og hydrogen, negative utslipp og lavere bruk av energi", "C. Redusert plantevekst", "D. Økt nedbør"],
        "answer": "B",
        "explanation": "For å håndtere klimaendringene må vi få på plass fornybar energi, kraft fra elektrisitet og hydrogen, negative utslipp og lavere bruk av energi i løpet av de neste 50 årene."
    },
    {
        "question": "Hva kan hjelpe klimaet ved å oppfylle FNs bærekraftsmål?",
        "options": ["A. Økt bruk av fossile brensler", "B. Økt nedbør", "C. Mat, utdanning og sikkerhet til alle, likestilling og bærekraftig produksjon av varer", "D. Redusert plantevekst"],
        "answer": "C",
        "explanation": "Å oppfylle FNs bærekraftsmål om mat, utdanning og sikkerhet til alle, likestilling og bærekraftig produksjon av varer kan hjelpe klimaet."
    },
    {
        "question": "Hva er viktig å lære seg for å håndtere vanskelige valg om klimaløsninger?",
        "options": ["A. Å øke nedbøren", "B. Å redusere planteveksten", "C. Å snakke om vanskelige valg uten å gå i vranglås", "D. Å øke solinnstrålingen"],
        "answer": "C",
        "explanation": "Det er viktig å lære seg å snakke om vanskelige valg om klimaløsninger uten å gå i vranglås for å håndtere utfordringene."
    },
    {
        "question": "Hva er hovedårsaken til dagens klimaendringer?",
        "options": ["A. Naturlige variasjoner i solens aktivitet", "B. Vulkanutbrudd", "C. Økt drivhuseffekt på grunn av menneskelige aktiviteter", "D. Endringer i jordens bane"],
        "answer": "C",
        "explanation": "Hovedårsaken til dagens klimaendringer er økt konsentrasjon av drivhusgasser som CO2, metan og lystgass, hovedsakelig fra forbrenning av fossile brensler."
    },
    {
        "question": "Hvor mye har gjennomsnittstemperaturen på jorden økt de siste 150 årene?",
        "options": ["A. 0,5 grader Celsius", "B. 1 grad Celsius", "C. 2 grader Celsius", "D. 3 grader Celsius"],
        "answer": "B",
        "explanation": "Gjennomsnittstemperaturen på jorden har økt med omtrent 1 grad Celsius siden den industrielle revolusjonen."
    },
    {
        "question": "Hva skjer med havnivået når havvannet varmes opp?",
        "options": ["A. Det synker", "B. Det forblir uendret", "C. Det stiger", "D. Det varierer tilfeldig"],
        "answer": "C",
        "explanation": "Når havvannet varmes opp, utvider det seg, noe som bidrar til stigende havnivå."
    },
    {
        "question": "Hva er en av konsekvensene av varmere luft?",
        "options": ["A. Mindre fordampning", "B. Mer fordampning", "C. Stabilt nedbørsmønster", "D. Mindre ekstremvær"],
        "answer": "B",
        "explanation": "Varmere luft fører til mer fordampning, noe som kan resultere i mer nedbør og endrede nedbørsmønstre."
    },
    {
        "question": "Hva er en av de største kildene til drivhusgasser?",
        "options": ["A. Fornybar energi", "B. Kjernekraft", "C. Fossile brensler", "D. Vannkraft"],
        "correct_answer": "C. Fossile brensler",
        "explanation": "Forbrenning av fossile brensler som olje, kull og gass er en av de største kildene til drivhusgasser."
    },
    {
        "question": "Hva er et mulig resultat av smeltende permafrost?",
        "options": ["A. Redusert metanutslipp", "B. Økt metanutslipp", "C. Stabilt klima", "D. Mindre skogbranner"],
        "answer": "B",
        "explanation": "Smeltende permafrost kan frigjøre metan, en kraftig drivhusgass, som har vært fanget i isen."
    },
    {
        "question": "Hva er et 'vippepunkt' i klimaforskningen?",
        "options": ["A. Et punkt der klimaet stabiliserer seg", "B. Et punkt der små endringer kan føre til store og irreversible endringer", "C. Et punkt der utslippene stopper", "D. Et punkt der temperaturen synker"],
        "answer": "B",
        "explanation": "'Vippepunkter' refererer til kritiske terskler der små endringer i klimaet kan føre til store og potensielt irreversible endringer."
    },
    {
        "question": "Hva er målet for global oppvarming som ble vedtatt i Parisavtalen?",
        "options": ["A. Maksimalt 1 grad Celsius over førindustrielt nivå", "B. Maksimalt 1,5 grader Celsius over førindustrielt nivå", "C. Maksimalt 2 grader Celsius over førindustrielt nivå", "D. Maksimalt 3 grader Celsius over førindustrielt nivå"],
        "answer": "C",
        "explanation": "'Parisavtalen' har som mål å begrense den globale oppvarmingen til godt under 2 grader Celsius over førindustrielt nivå."
    },
    {
        "question": "Hva er en av de viktigste tiltakene for å redusere utslipp av drivhusgasser?",
        "options": ["A. Økt bruk av fossile brensler", "B. Redusert bruk av fornybar energi", "C. Økt energieffektivitet og bruk av fornybare energikilder", "D. Økt avskoging"],
        "answer": "C",
        "explanation": "For å redusere utslipp av drivhusgasser er det viktig å øke energieffektiviteten og bruke mer fornybare energikilder som sol, vind og vannkraft."
    },
    {
        "question": "Hva er en av konsekvensene av økt ekstremvær?",
        "options": ["A. Mindre flom og jordras", "B. Mer flom og jordras", "C. Stabilt værmønster", "D. Mindre skader på infrastruktur"],
        "answer": "B",
        "explanation": "Økt ekstremvær kan føre til mer flom og jordras, noe som kan skade hus, byer og landbruk."
    }

]

# Initialiser GUI
root = tk.Tk()
root.title("Quiz App")
root.geometry("600x550")
style = Style(theme="flatly")
style.configure("TLabel", font=("Helvetica", 16))
style.configure("TButton", font=("Helvetica", 14))

# Globale variabler
score = 0
current_question = 0
choice_btns = []
quiz_data = []
user_answers = []

# Widgets
qs_label = ttk.Label(root, anchor="center", wraplength=500, padding=10)
qs_label.pack(pady=10)

choices_frame = ttk.Frame(root)
choices_frame.pack(pady=10, fill=tk.X)

feedback_label = ttk.Label(root, anchor="center", padding=10)
feedback_label.pack(pady=10)

score_label = ttk.Label(root, text="", anchor="center", padding=10)
score_label.pack(pady=10)

nav_frame = ttk.Frame(root)
nav_frame.pack(pady=10)

prev_btn = ttk.Button(nav_frame, text="Tilbake", command=lambda: previous_question(), state="disabled")
prev_btn.pack(side="left", padx=10)

next_btn = ttk.Button(nav_frame, text="Neste", command=lambda: next_question(), state="disabled")
next_btn.pack(side="left", padx=10)

retry_btn = ttk.Button(root, text="Prøv igjen", command=lambda: restart_quiz(), state="disabled")
retry_btn.pack(pady=10)

progress = ttk.Progressbar(root, orient="horizontal", length=400, mode="determinate")
progress.pack(pady=10)

# --- Funksjoner ---

def restart_quiz():
    global score, current_question, choice_btns, quiz_data, user_answers
    score = 0
    current_question = 0
    quiz_data = original_quiz_data.copy()
    random.shuffle(quiz_data)
    user_answers = [None] * len(quiz_data)
    progress.config(maximum=len(quiz_data), value=0)
    retry_btn.config(state="disabled")
    next_btn.config(state="disabled")
    prev_btn.config(state="disabled")
    show_question()

def show_question():
    global choice_btns
    for btn in choice_btns:
        btn.destroy()
    choice_btns = []

    question = quiz_data[current_question]
    qs_label.config(text=question["question"])
    options = question["options"]

    selected_index = user_answers[current_question]

    for i, option in enumerate(options):
        button = ttk.Button(
            choices_frame,
            text=option,
            command=lambda i=i: check_answer(i),
            style="TButton"
        )
        button.pack(pady=5, fill=tk.X)
        choice_btns.append(button)

    feedback_label.config(text="")
    score_label.config(text="Score: {}/{}".format(score, len(quiz_data)))
    prev_btn.config(state="normal" if current_question > 0 else "disabled")
    next_btn.config(state="disabled")
    retry_btn.config(state="disabled")

    if selected_index is not None:
        apply_feedback_colors(selected_index)
        next_btn.config(state="normal")

def check_answer(choice_index):
    global score
    question = quiz_data[current_question]
    selected_letter = chr(65 + choice_index)
    correct_letter = question["answer"]
    correct_index = ord(correct_letter) - 65
    previous = user_answers[current_question]

    # Juster score hvis bytter svar
    if previous is not None and chr(65 + previous) == correct_letter:
        score -= 1

    user_answers[current_question] = choice_index

    if selected_letter == correct_letter:
        score += 1
        feedback_label.config(text="Riktig! " + question['explanation'], foreground="green")
    else:
        feedback_label.config(
            text=f"Galt! Det riktige svaret er {correct_letter}. {question['explanation']}",
            foreground="red"
        )

    score_label.config(text="Score: {}/{}".format(score, len(quiz_data)))
    progress['value'] = current_question + 1

    apply_feedback_colors(choice_index)
    next_btn.config(state="normal")

def apply_feedback_colors(selected_index):
    question = quiz_data[current_question]
    correct_index = ord(question["answer"]) - 65

    for i, button in enumerate(choice_btns):
        button.config(state="disabled")
        if i == correct_index:
            button.config(style="success.TButton")
        elif i == selected_index:
            button.config(style="danger.TButton")
        else:
            button.config(style="secondary.TButton")

def next_question():
    global current_question
    if current_question < len(quiz_data) - 1:
        current_question += 1
        show_question()
    else:
        messagebox.showinfo(
            "Quiz gjennomført",
            f"Du fikk {score} av {len(quiz_data)} riktige.\nTakk for at du deltok!"
        )
        retry_btn.config(state="normal")
        next_btn.config(state="disabled")
        prev_btn.config(state="disabled")

def previous_question():
    global current_question
    if current_question > 0:
        current_question -= 1
        show_question()

# Start første gang
restart_quiz()
root.mainloop()
```
