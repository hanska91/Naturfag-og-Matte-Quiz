```python
!pip install ttkbootstrap
import tkinter as tk
from tkinter import messagebox, ttk
from ttkbootstrap import Style
import random

# Quiz data
quiz_data = [
    {
        "question": "Hva er hovedforskjellen mellom tradisjonell og moderne bioteknologi?",
        "options": ["1. Tradisjonell bioteknologi bruker naturlige prosesser, mens moderne bioteknologi bruker genteknologi", "2. Tradisjonell bioteknologi er mer effektiv", "3. Moderne bioteknologi er kun brukt i medisin", "4. Tradisjonell bioteknologi er dyrere"],
        "answer": 1,
        "explanation": "Tradisjonell bioteknologi bruker naturlige prosesser som fermentering, mens moderne bioteknologi inkluderer genteknologi og molekylærbiologi."
    },
    {
        "question": "Hva er kloning?",
        "options": ["1. Å lage genetiske kopier av organismer", "2. Å redigere gener i en organisme", "3. Å lage kunstige organer", "4. Å øke veksthastigheten til planter"],
        "answer": 1,
        "explanation": "Kloning er prosessen med å lage genetiske kopier av organismer."
    },
    {
        "question": "Hva er et etisk dilemma ved bruk av kloning?",
        "options": ["1. Økt matproduksjon", "2. Potensielle helsefarer", "3. Bedre smak", "4. Redusert genetisk mangfold"],
        "answer": 4,
        "explanation": "Et etisk dilemma ved bruk av kloning er redusert genetisk mangfold, som kan påvirke økosystemer og arter."
    },
    {
        "question": "Hva kan stamceller brukes til?",
        "options": ["1. Å bekjempe infeksjoner", "2. Å lage ulike celletyper", "3. Å reparere DNA", "4. Å forhindre aldring"],
        "answer": 2,
        "explanation": "Stamceller kan utvikle seg til ulike celletyper og brukes i forskning på sykdomsbehandling."
    },
    {
        "question": "Hva er et etisk dilemma ved bruk av stamceller?",
        "options": ["1. Økt matproduksjon", "2. Potensielle helsefarer", "3. Bruk av embryonale stamceller", "4. Redusert bruk av sprøytemidler"],
        "answer": 3,
        "explanation": "Et etisk dilemma ved bruk av stamceller er bruken av embryonale stamceller, som reiser spørsmål om livets begynnelse og rettigheter."
    },
    {
        "question": "Hva betyr GMO?",
        "options": ["1. Genmodifiserte organismer", "2. Genetisk mikset organismer", "3. Gener med overføring", "4. Genetisk moderne organismer"],
        "answer": 1,
        "explanation": "GMO står for genmodifiserte organismer, som betyr organismer med endret arvemateriale."
    },
    {
        "question": "Hva er CRISPR-teknologi mest kjent for?",
        "options": ["1. Å klone organismer", "2. Å klippe og lime DNA presist", "3. Å øke veksthastigheten til planter", "4. Å lage kunstige organer"],
        "answer": 2,
        "explanation": "CRISPR brukes til å redigere DNA på en presis og effektiv måte."
    },
    {
        "question": "Hva er et etisk dilemma ved bruk av GMO?",
        "options": ["1. Økt matproduksjon", "2. Potensielle helsefarer", "3. Bedre smak", "4. Redusert bruk av sprøytemidler"],
        "answer": 2,
        "explanation": "Et etisk dilemma ved bruk av GMO er potensielle helsefarer og miljøpåvirkninger."
    },
    {
        "question": "Hva brukes DNA-profiler til?",
        "options": ["1. Å bestemme kjønn", "2. Å identifisere personer", "3. Å forbedre synet", "4. Å lage nye organismer"],
        "answer": 2,
        "explanation": "DNA-profiler brukes for å identifisere personer, for eksempel i rettsmedisin."
    },
    {
        "question": "Hva er genterapi?",
        "options": ["1. Å klone organismer", "2. Å redigere gener for å behandle sykdommer", "3. Å lage kunstige organer", "4. Å øke veksthastigheten til planter"],
        "answer": 2,
        "explanation": "Genterapi innebærer å redigere gener for å behandle eller forebygge sykdommer."
    },
    {
        "question": "Hva er et etisk dilemma ved bruk av genteknologi?",
        "options": ["1. Økt matproduksjon", "2. Muligheten for designerbabyer", "3. Bedre smak", "4. Redusert bruk av sprøytemidler"],
        "answer": 2,
        "explanation": "Muligheten for å skape 'designerbabyer' er et etisk dilemma knyttet til genteknologi."
    },
    {
        "question": "Hva er et annet etisk dilemma ved bruk av genteknologi?",
        "options": ["1. Økt matproduksjon", "2. Potensielle helsefarer", "3. Bedre smak", "4. Redusert bruk av sprøytemidler"],
        "answer": 2,
        "explanation": "Et etisk dilemma ved bruk av genteknologi er potensielle helsefarer og miljøpåvirkninger."
    },
    {
        "question": "Hva er avl?",
        "options": ["1. Å redigere gener i en organisme", "2. Å lage genetiske kopier av organismer", "3. Å selektere og pare organismer for ønskede egenskaper", "4. Å lage kunstige organer"],
        "answer": 3,
        "explanation": "Avl er prosessen med å selektere og pare organismer for å oppnå ønskede egenskaper."
    },
    {
        "question": "Hvilken risiko er knyttet til genterapi?",
        "options": ["1. Allergiske reaksjoner", "2. Utilsiktede genetiske endringer", "3. Økt metabolisme", "4. Forbedret immunforsvar"],
        "answer": 2,
        "explanation": "En risiko ved genterapi er utilsiktede genetiske endringer som kan føre til komplikasjoner."
    },
    {
        "question": "Hva er et etisk dilemma ved bruk av genterapi?",
        "options": ["1. Økt matproduksjon", "2. Potensielle helsefarer", "3. Bedre smak", "4. Uforutsigbare langtidseffekter"],
        "answer": 4,
        "explanation": "Et etisk dilemma ved bruk av genterapi er uforutsigbare langtidseffekter på mennesker og fremtidige generasjoner."
    },
    {
        "question": "Hva er etisk kontroversielt med gentester?",
        "options": ["1. De er unøyaktige", "2. Personvern og diskriminering", "3. De er for dyre", "4. De tar for lang tid"],
        "answer": 2,
        "explanation": "Gentester kan være etisk kontroversielle på grunn av personvern og risiko for diskriminering basert på genetisk informasjon."
    },
    {
        "question": "Hva er DNA-sekvensering?",
        "options": ["1. En metode for å klone DNA", "2. En teknikk for å bestemme rekkefølgen av nukleotider i DNA", "3. En prosess for å lage genetiske kopier av organismer", "4. En metode for å redigere gener"],
        "answer": 2,
        "explanation": "DNA-sekvensering er en teknikk som brukes for å bestemme rekkefølgen av nukleotider i et DNA-molekyl."
    },
    {
        "question": "Hva står PGD for, og hva brukes det til?",
        "options": ["1. Preimplantasjon genetisk diagnostikk, brukt for å identifisere genetiske defekter i embryoer før implantasjon", "2. Postnatal genetisk diagnostikk, brukt for å identifisere genetiske defekter etter fødsel", "3. Prenatal genetisk diagnostikk, brukt for å identifisere genetiske defekter under graviditet", "4. Preimplantasjon genetisk diagnostikk, brukt for å klone embryoer"],
        "answer": 1,
        "explanation": "PGD står for preimplantasjon genetisk diagnostikk og brukes for å identifisere genetiske defekter i embryoer før de implanteres i livmoren."
    },
    {
        "question": "Hva er NIPT?",
        "options": ["1. En metode for å redigere gener", "2. En ikke-invasiv prenatal test for å oppdage kromosomavvik hos fosteret", "3. En teknikk for å klone embryoer", "4. En metode for å lage genetiske kopier av organismer"],
        "answer": 2,
        "explanation": "NIPT står for ikke-invasiv prenatal test og brukes for å oppdage kromosomavvik hos fosteret ved å analysere fosterets DNA i morens blod."
    },
    {
        "question": "Hva er Huntingtons sykdom?",
        "options": ["1. En genetisk sykdom som påvirker muskelstyrken", "2. En arvelig nevrodegenerativ sykdom som påvirker hjernen", "3. En infeksjonssykdom som påvirker lungene", "4. En genetisk sykdom som påvirker synet"],
        "answer": 2,
        "explanation": "Huntingtons sykdom er en arvelig nevrodegenerativ sykdom som fører til progressiv skade på hjerneceller."
    },
    {
        "question": "Hva er Duchennes muskeldystrofi?",
        "options": ["1. En genetisk sykdom som påvirker synet", "2. En arvelig muskelsykdom som fører til progressiv muskelsvakhet", "3. En infeksjonssykdom som påvirker lungene", "4. En genetisk sykdom som påvirker hjernen"],
        "answer": 2,
        "explanation": "Duchennes muskeldystrofi er en arvelig muskelsykdom som fører til progressiv muskelsvakhet og tap av muskelmasse."
    },
    {
        "question": "Hva er Golden Rice?",
        "options": ["1. En type ris som er genetisk modifisert for å inneholde vitamin A", "2. En type ris som vokser raskere enn vanlig ris", "3. En type ris som er resistent mot skadedyr", "4. En type ris som er genetisk modifisert for å inneholde jern"],
        "answer": 1,
        "explanation": "Golden Rice er en type ris som er genetisk modifisert for å inneholde beta-karoten, som kroppen omdanner til vitamin A."
    },
    {
        "question": "Hva er biodrivstoff?",
        "options": ["1. Drivstoff laget av fossile brensler", "2. Drivstoff laget av fornybare biologiske materialer", "3. Drivstoff laget av kjemiske forbindelser", "4. Drivstoff laget av genetisk modifiserte organismer"],
        "answer": 2,
        "explanation": "Biodrivstoff er drivstoff laget av fornybare biologiske materialer som planter og alger."
    },
    {
        "question": "Hvordan har CRISPR-teknologi bidratt til å redusere bruk av sprøytemidler mot soppen meldugg?",
        "options": ["1. Ved å klone planter som er resistente mot meldugg", "2. Ved å redigere gener i planter for å gjøre dem resistente mot meldugg", "3. Ved å lage kunstige sprøytemidler", "4. Ved å øke veksthastigheten til planter"],
        "answer": 2,
        "explanation": "CRISPR-teknologi brukes til å redigere gener i planter for å gjøre dem resistente mot soppen meldugg, noe som reduserer behovet for sprøytemidler."
    },
    {
        "question": "Hva er et etisk dilemma ved bruk av GMO-griser som er motstandsdyktige mot virus som gir svinepest?",
        "options": ["1. Økt matproduksjon", "2. Potensielle helsefarer", "3. Bedre smak", "4. Redusert bruk av sprøytemidler"],
        "answer": 2,
        "explanation": "Et etisk dilemma ved bruk av GMO-griser er potensielle helsefarer og miljøpåvirkninger."
    },
    {
        "question": "Hva er CRISPR-CAS9 proteiner kjent for?",
        "options": ["1. Å klone organismer", "2. Å klippe og lime DNA presist", "3. Å øke veksthastigheten til planter", "4. Å lage kunstige organer"],
        "answer": 2,
        "explanation": "CRISPR-CAS9 proteiner brukes til å redigere DNA på en presis og effektiv måte."
    },
    {
        "question": "Hva gjør enzymer som bryter ned plast?",
        "options": ["1. De redigerer gener i bakterier", "2. De kloner bakterier", "3. De bryter ned plast til mindre molekyler", "4. De lager kunstige organer"],
        "answer": 3,
        "explanation": "Enzymer som bryter ned plast, katalyserer nedbrytningen av plast til mindre molekyler som kan brytes ned videre av bakterier."
    },
    {
        "question": "Hva er genmodifisering av bakterienes plasmider?",
        "options": ["1. Å lage genetiske kopier av bakterier", "2. Å redigere gener i bakterienes plasmider for å gi dem nye egenskaper", "3. Å øke veksthastigheten til bakterier", "4. Å lage kunstige organer"],
        "answer": 2,
        "explanation": "Genmodifisering av bakterienes plasmider innebærer å redigere gener i plasmidene for å gi bakteriene nye egenskaper, som for eksempel produksjon av medisiner."
    },
    {
        "question": "Hvor stor andel av soya som produseres er genmodifisert?",
        "options": ["1. 50 %", "2. 60 %", "3. 70 %", "4. 80 %"],
        "answer": 4,
        "explanation": "Omtrent 80 % av soya som produseres er genmodifisert for å tåle ugressmidler eller for å forbedre avlingene."
    },
    {
        "question": "Hvor dyrkes 90 % av GMO-planter?",
        "options": ["1. USA, Brasil, Argentina, India og Canada", "2. Kina, Russland, Tyskland, Frankrike og Japan", "3. Australia, New Zealand, Sør-Afrika, Egypt og Mexico", "4. Storbritannia, Italia, Spania, Portugal og Nederland"],
        "answer": 1,
        "explanation": "90 % av arealet det dyrkes GMO-planter er i USA, Brasil, Argentina, India og Canada."
    },
    {
        "question": "Hva er bioteknologiloven og genteknologiloven i Norge?",
        "options": ["1. Lover som regulerer bruk av bioteknologi og genteknologi", "2. Lover som forbyr all bruk av bioteknologi", "3. Lover som tillater ubegrenset bruk av genteknologi", "4. Lover som regulerer import av bioteknologiske produkter"],
        "answer": 1,
        "explanation": "Bioteknologiloven og genteknologiloven i Norge er lover som regulerer bruk av bioteknologi og genteknologi for å sikre etisk og sikker bruk."
    },
    {
        "question": "Hva er mais, oljeraps og soya genmodifisert for å tåle?",
        "options": ["1. Soppinfeksjoner", "2. Ugressmiddelet glyfosfat", "3. Tørke", "4. Skadedyr"],
        "answer": 2,
        "explanation": "Mais, oljeraps og soya er genmodifisert for å tåle ugressmiddelet glyfosfat, som ellers ville ødelagt avlingene."
    },
    {
        "question": "Hva er GMO-mikroorganismers rolle i medisinsk industri?",
        "options": ["1. Å produsere insulin", "2. Å lage plastmaterialer", "3. Å bryte ned plastavfall", "4. Å øke plantevekst"],
        "answer": 1,
        "explanation": "GMO-mikroorganismer, som genmodifiserte bakterier, brukes til å produsere medisiner som insulin til behandling av diabetes."
    },
    {
        "question": "Hvilket land har strenge lover som bioteknologiloven og genteknologiloven?",
        "options": ["1. USA", "2. Norge", "3. Kina", "4. India"],
        "answer": 2,
        "explanation": "Norge har strenge lover som bioteknologiloven og genteknologiloven, som regulerer bruken av bioteknologi og genteknologi."
    },
    {
        "question": "Hva er et mulig etisk dilemma med genetisk modifisering av planter for å tåle sprøytemidler?",
        "options": ["1. Økt matproduksjon", "2. Potensielle miljøkonsekvenser", "3. Bedre smak", "4. Økte kostnader for bønder"],
        "answer": 2,
        "explanation": "Et etisk dilemma er de potensielle miljøkonsekvensene, som for eksempel tap av biologisk mangfold eller uforutsette miljøpåvirkninger."
    },
    {
        "question": "Hvordan kan CRISPR-teknologi bidra til bærekraftig matproduksjon?",
        "options": ["1. Ved å gjøre planter mer motstandsdyktige mot sykdommer", "2. Ved å øke avlingene gjennom kunstig befruktning", "3. Ved å bruke færre naturressurser", "4. Ved å gjøre dyr mer produktive"],
        "answer": 1,
        "explanation": "CRISPR-teknologi kan gjøre planter mer motstandsdyktige mot sykdommer og skadedyr, noe som kan redusere behovet for sprøytemidler."
    },
    {
        "question": "Hva er et mulig problem ved bruk av genmodifiserte organismer (GMO) i landbruket?",
        "options": ["1. Økt avkastning på grunn av redusert skadedyrbekjempelse", "2. Risiko for uforutsette effekter på økosystemet", "3. Økt pris på matvarer", "4. Økt biologisk mangfold"],
        "answer": 2,
        "explanation": "Det er bekymring for at GMO kan ha uforutsette effekter på økosystemet, som for eksempel å påvirke andre arter eller redusere biologisk mangfold."
    },
    {
        "question": "Hvordan fungerer CRISPR-Cas9-teknologien i genmodifisering?",
        "options": ["1. Den klipper DNA-strengen på spesifikke steder", "2. Den kopierer genetiske sekvenser", "3. Den kloner organismer", "4. Den reparerer skadet DNA tilfeldig"],
        "answer": 1,
        "explanation": "CRISPR-Cas9-teknologien fungerer ved å klippe DNA-strengen på spesifikke steder, noe som gjør det mulig å endre eller redigere gener."
    },
    {
        "question": "Hva er hovedformålet med Golden Rice?",
        "options": ["1. Å øke vitamin A-nivåene i ris", "2. Å gjøre ris mer motstandsdyktig mot sykdommer", "3. Å øke avlingen av ris", "4. Å redusere bruken av sprøytemidler i risdyrking"],
        "answer": 1,
        "explanation": "Golden Rice er genmodifisert for å inneholde beta-karoten, som kroppen kan omdanne til vitamin A, og dermed bidra til å redusere vitamin A-mangel."
    },
    {
        "question": "Hva er et eksempel på bruk av genmodifiserte bakterier i helseindustrien?",
        "options": ["1. Produksjon av medisiner som feks. insulin", "2. Produksjon av plastmaterialer", "3. Produksjon av drivstoff", "4. Produksjon av mat"], 
        "answer": 1,
        "explanation": "Genmodifiserte bakterier brukes til å produsere medisiner, som for eksempel insulin for behandling av diabetes."
    },
    {
        "question": "Hva regulerer genteknologiloven i Norge?",
        "options": ["1. Bruken av genteknologi i medisin og forskning", "2. Bruken av bioteknologi i landbruket", "3. Bruken av genmodifiserte organismer i matproduksjon", "4. Bruken av genteknologi i behandling av mennesker og dyr"],
        "answer": 1,
        "explanation": "Genteknologiloven regulerer bruken av genteknologi i medisin, forskning og behandling av mennesker og dyr, for å sikre etisk og sikker praksis."
    },
    {
        "question": "Hva er hovedformålet med bioteknologiloven?",
        "options": ["1. Å beskytte mennesker mot uønskede genetiske endringer", "2. Å regulere bruken av bioteknologi i helsevesenet", "3. Å sikre sikker og etisk bruk av bioteknologi", "4. Å tillate ubegrenset bruk av bioteknologi for medisinske formål"],
        "answer": 3,
        "explanation": "Bioteknologiloven har som hovedformål å sikre etisk og sikker bruk av bioteknologi i Norge, og regulere ulike bioteknologiske metoder og produkter."
    },
    {
        "question": "Hvilket av følgende tiltak er regulert av genteknologiloven i Norge?",
        "options": ["1. Kloning av mennesker", "2. Genmodifisering av planter for matproduksjon", "3. Bruk av genmodifiserte bakterier i industrien", "4. Bruk av genteknologi i medisinsk behandling av mennesker"],
        "answer": 4,
        "explanation": "Genteknologiloven regulerer bruken av genteknologi i medisinsk behandling av mennesker, som for eksempel i behandlingen av genetiske sykdommer."
    },
    {
        "question": "Hva er en viktig etisk bekymring som bioteknologiloven i Norge tar hensyn til?",
        "options": ["1. Redusere matproduksjon", "2. Beskytte mennesker mot genetisk diskriminering", "3. Øke landbruksproduksjon med GMO-planter", "4. Øke bruken av genetiske tester for alle"],
        "answer": 2,
        "explanation": "En viktig etisk bekymring bioteknologiloven tar hensyn til, er å beskytte mennesker mot genetisk diskriminering og misbruk av genetisk informasjon."
    },
    {
        "question": "Hvilket av følgende er et formål med bioteknologiloven?",
        "options": ["1. Å tillate fri bruk av genmodifiserte organismer uten regulering", "2. Å regulere og kontrollere bruk av genmodifiserte organismer i matproduksjon og landbruk", "3. Å gjøre genmodifisering av mennesker lovlig", "4. Å forby alle former for genteknologi"],
        "answer": 2,
        "explanation": "Bioteknologiloven regulerer og kontrollerer bruken av genmodifiserte organismer i matproduksjon og landbruk, med tanke på etikk og sikkerhet."
    },
    {
        "question": "Hvem overvåker og håndhever genteknologiloven i Norge?",
        "options": ["1. Folkehelseinstituttet", "2. Mattilsynet", "3. Norges forskningsråd", "4. Bioteknologirådet"],
        "answer": 4,
        "explanation": "Bioteknologirådet er et statlig organ som gir råd om etiske og juridiske spørsmål knyttet til bioteknologi og håndhever genteknologiloven i Norge."
    },
    {
        "question": "Hva skjer dersom en genetisk modifisert organisme ikke er i tråd med bioteknologiloven?",
        "options": ["1. Den får automatisk godkjenning til bruk", "2. Den kan bli forbudt i Norge", "3. Den får et midlertidig unntak for bruk", "4. Den blir brukt i forskning før videre godkjenning"],
        "answer": 2,
        "explanation": "Dersom en genetisk modifisert organisme ikke er i tråd med bioteknologiloven, kan den bli forbudt i Norge for å beskytte helse, miljø og etikk."
    }

]

# Shuffle questions for randomness
random.shuffle(quiz_data)

# Create the main window
root = tk.Tk()
root.title("Quiz App")
root.geometry("600x500")
style = Style(theme="flatly")

# Configure the font size for the question and choice buttons
style.configure("TLabel", font=("Helvetica", 16))
style.configure("TButton", font=("Helvetica", 14))

# Initialize the score
score = 0
current_question = 0

# Create the question label
qs_label = ttk.Label(
    root,
    anchor="center",
    wraplength=500,
    padding=10
)
qs_label.pack(pady=10)

# Create the choice buttons
choice_btns = []
for i in range(4):
    button = ttk.Button(
        root,
        command=lambda i=i: check_answer(i)
    )
    button.pack(pady=5)
    choice_btns.append(button)

# Create the feedback label
feedback_label = ttk.Label(
    root,
    anchor="center",
    padding=10
)
feedback_label.pack(pady=10)

# Create the score label
score_label = ttk.Label(
    root,
    text="Score: 0/{}".format(len(quiz_data)),
    anchor="center",
    padding=10
)
score_label.pack(pady=10)

# Create the next button
next_btn = ttk.Button(
    root,
    text="Next",
    command=lambda: next_question(),
    state="disabled"
)
next_btn.pack(pady=10)

# Function to display the current question and choices
def show_question():
    question = quiz_data[current_question]
    qs_label.config(text=question["question"])

    # Display the choices on the buttons
    options = question["options"]
    for i in range(4):
        choice_btns[i].config(text=options[i], state="normal")

    # Clear the feedback label and disable the next button
    feedback_label.config(text="")
    next_btn.config(state="disabled")

# Function to check the selected answer and provide feedback
def check_answer(choice):
    global score
    question = quiz_data[current_question]
    selected_choice = choice_btns[choice].cget("text")[0]  # Get the letter (A, B, C, D)

    # Check if the selected choice matches the correct answer
    if selected_choice == question["answer"]:
        score += 1
        feedback_label.config(text="Correct! " + question['explanation'], foreground="green")
    else:
        feedback_label.config(
            text=f"Incorrect! The correct answer is {question['answer']}. {question['explanation']}",
            foreground="red"
        )

    score_label.config(text="Score: {}/{}".format(score, len(quiz_data)))

    # Disable all choice buttons and enable the next button
    for button in choice_btns:
        button.config(state="disabled")
    next_btn.config(state="normal")

# Function to move to the next question
def next_question():
    global current_question
    current_question += 1

    if current_question < len(quiz_data):
        show_question()
    else:
        # Custom message based on score
        if score == len(quiz_data):
            msg = "Excellent! You got all the questions right!"
        elif score >= len(quiz_data) // 2:
            msg = "Good job! You scored {}/{}.".format(score, len(quiz_data))
        else:
            msg = "Better luck next time! You scored {}/{}.".format(score, len(quiz_data))

        messagebox.showinfo("Quiz Completed", msg)
        root.destroy()

# Show the first question
show_question()

# Start the main event loop
root.mainloop()
```
